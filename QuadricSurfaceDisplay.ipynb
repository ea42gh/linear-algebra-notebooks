```python
import numpy as np
import holoviews as hv

hv.extension('plotly')
```

```python
N = 100
x = np.linspace(-4, 4,N); y = np.linspace(-2.5,2.5,N)

xx,yy   = np.meshgrid(x,y)

def show_plot( f, zcut, lvl, cmap="spectral"):
    zz    = f(xx,yy)
    plane = zcut(lvl,xx,yy)
    surface = hv.Surface(zz).opts(cmap=cmap,height=600, width=800)
    return surface*hv.Surface(plane )
```

Consider the function
$$
\quad\quad f(x,y) = x^2+ 2 x y + 3 y^2
$$.

We will plot $z = f(x,y)$ and a cross section $z = 10$, i.e., the function $x^2+ 2 x y + 3 y^2 = 10$.

```python
def f(x,y):         return x*x + 3*y*y + 2*x*y
def zcut(z,xx,yy):  return z+0*xx

hv.DynamicMap( lambda level : show_plot(f, zcut, level, cmap="rainbow"), kdims="level" ).redim.range(level=(0.,10))
```

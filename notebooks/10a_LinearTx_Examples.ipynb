# 1. Review: Basic Definitions and Theorems

## 1.1 Definition

<div style="float:left;background-color:#F2F5A9;color:black;padding:2mm 2mm 2mm 2mm;">

**Definition:** A function $T : U \rightarrow V$ over scalars $\mathbb{F}$<br>$\quad$ is a **linear transformation** from a vector space $U$ to a vector space $V$ iff
* $\forall u_1, u_2 \in U \quad\quad\;\; T \;(u_1+u_1) = T u_1 + T u_2$, and
* $\forall u \in U, \forall \alpha \in \mathbb{F} \quad T\ (\alpha u) = \alpha T u$
</div>
<div style="float:left;padding-left:1cm;">

**Remark:**<br> 
$\quad$ The two conditions in this definition can be combined:<br><br>
$$
\boxed{ \forall u_1, u_2 \in U, \; \forall \alpha,\beta \in \mathbb{F} \quad T\ (\alpha\ u_1 + \beta\ u_2) = \alpha\ T u_1 + \beta\ T u_2 }
$$

</div>

**Remarks:** 
* $T$ maps a kplane in $U$ onto a kplane in $V$.
* $T$ transforms parallelograms into parallelograms

---
The **main example** of a linear transformation is $y = A x$, where the vectors $x \in \mathbb{F}^N, y \in \mathbb{F}^M$, and the matrix $A \in \mathbb{F}^{M \times N}$.

<strong style="color:darkred;font-size:20px;">Important Property: Linear Transformations can be represented by a matrix:<strong>

<div style="float:left;background-color:#F2F5A9;color:black;padding:2mm 2mm 2mm 2mm;height:4cm;width:46%;">

**Theorem:** Let $T: \mathbb{F}^N \rightarrow \mathbb{F}^M$ be a linear transformation,<br>
$\qquad$ and let $e_i$ be the $i^{th}$ column of $I_N$.<br>
$\qquad$ Then $\quad$ $T x = A_T x,$<br><br>
$\qquad$ where $A_T$ is given by
    $$A_T = \left( T e_1 \mid T e_2 \mid \dots \mid T e_N \right)$$
</div>
<div style = "float:left;padding-left:1cm;">

**Example:**<br>
$$
\left.
\begin{align}
\begin{pmatrix} 1\\0\end{pmatrix} & \xrightarrow{T} \begin{pmatrix}2\\3\end{pmatrix} \\
\begin{pmatrix} 0\\1\end{pmatrix} & \xrightarrow{T} \begin{pmatrix}4\\1\end{pmatrix}
\end{align}
\; \right\}
\;\Leftrightarrow \;
$$
</div>
<div style = "float:left;padding-top:0.0cm;">

$
\left\{ \; \begin{align}
T \begin{pmatrix} x \\ y \end{pmatrix} &= T \left( x \begin{pmatrix} 1\\0\end{pmatrix} + y \begin{pmatrix} 0\\1\end{pmatrix} \right) \\
                                       &=          x \begin{pmatrix} 2\\3\end{pmatrix} + y \begin{pmatrix} 4\\1\end{pmatrix}  \\
                                       &= \begin{pmatrix} 2 & 4 \\ 3 & 1 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix}
\end{align}
\right.
$
</div>

## 1.2 One-to-One and ONTO Transformations

<div style="float:left;background-color:#F2F5A9;color:black;padding:2mm 2mm 2mm 2mm;height:3cm;width:46%;height:4cm;">

**Definition:** A function $f$ from a domain $U$ to a codomain $V$<br> $\quad f : U \rightarrow V$ is
* **ONTO** or **surjective** iff $f(u) = v$ has a solution for every $v \in V$.
* **1-1** or **injective** iff $\forall u_1, u_2 \in U \quad f(u_1) = f(u_2) \Rightarrow u_1 = u_2$
* **1-1 and ONTO** or **bijective** iff it is both injective and surjective.
</div>

<div style="float:right;background-color:#F2F5A9;color:black;padding:2mm 2mm 2mm 2mm;height:3cm;width:50%;height:4cm;">

**Theorem:** Given $y = A x$, where the vectors $x \in \mathbb{F}^N, y \in \mathbb{F}^M$, and<br> $\quad$ $A \in \mathbb{F}^{M \times N}$
* $y = A x$ is ONTO iff $A$ has a pivot in every row
* $y = A x$ is 1-1 iff $A$ has a pivot in every column

**Theorem:** A function is invertible iff it is both 1-1 and ONTO
</div>

**Remarks:**
* ONTO means that the codomain of the function is equal to the range of the function
* 1-1 means that any vector $v$ in the range the equation $f(u) = v$ has a unique solution.

---
**Example:**
* $y = A x$ with $A = \begin{pmatrix} \color{red}1 & 2 \\ 0 & \color{red}1 \end{pmatrix} \quad$ is both 1-1 and ONTO.
   It has inverse $A^{-1} = \begin{pmatrix} 1 & -2 \\ 0 & 1  \end{pmatrix}$

* $y = A x$ with $A = \begin{pmatrix} \color{red}1 & 2 & 3 \\ 0 & \color{red}2 & 3 \end{pmatrix} \quad$ is ONTO, but not 1-1

* $y = A x$ with $A = \begin{pmatrix} \color{red}1 & 2 \\ 0 & \color{red}2 \\ 0 & 0 \end{pmatrix} \quad$ is 1-1, but not ONTO

* $y = A x$ with $A = \begin{pmatrix} \color{red}1 & 2 \\ 0 & 0 \end{pmatrix} \quad$ is neither 1-1, nor ONTO

## 1.3 Composition of Linear Transformations

<div style="background-color:#F2F5A9;color:black;padding:2mm 2mm 2mm 2mm;width:60%;">

**Theorem:** The composition of linear transformations is linear.
</div>
<div style="width:60%;"><br>

**Notation:** let $S : U \rightarrow V$ and $T : V \rightarrow W$ be linear transformations, and let $u \in U$

$$ u \xrightarrow{\;\;v\ =\ S u\;\;} v \xrightarrow{\;\; w\ =\ T v\;\;} w \qquad \therefore \; w = T (S u) = T S u $$
</div>

**Example:**<br>
$\quad$ Let $\quad
S \;=\; \begin{pmatrix} 1 & 2 \\ 2 & 3 \end{pmatrix}, \;\; T \;=\; \begin{pmatrix} -1 & 1 \\ -2 & 1 \end{pmatrix}, \quad
$ then <br><br>
$\quad$ $T \left(\ S  \begin{pmatrix} x \\ y \end{pmatrix} \ \right)  \; = \;
\begin{pmatrix} -1 & 1 \\ -2 & 1 \end{pmatrix} \begin{pmatrix} 1 & 2 \\ 2 & 3 \end{pmatrix}
\begin{pmatrix} x \\ y \end{pmatrix} \;=\;
\begin{pmatrix} 1 & \; 1 \\ 0 & -1 \end{pmatrix}
\begin{pmatrix} x \\ y \end{pmatrix}$

**Remark:** the order matters! $\quad T S  \ne S T$<br>

$\quad$ $S\ T\  \begin{pmatrix} x \\ y \end{pmatrix}  \; = \;
 \begin{pmatrix} 1 & 2 \\ 2 & 3 \end{pmatrix} \begin{pmatrix} -1 & 1 \\ -2 & 1 \end{pmatrix}
\begin{pmatrix} x \\ y \end{pmatrix} \;=\;
\left( \begin{array}{rr} -5 & \; 3 \\ -8 & -5 \end{array} \right)
\begin{pmatrix} x \\ y \end{pmatrix}$

# 2. Matrix Representation of a Linear Transformation

## 2.1 Tools: Polar Coordinates and Congruent Triangles

Many **geometric transformations** are easily shown to be **linear**.

$\quad$ To obtain their matrix representations, we need to **find the coordinates of the transformed points.**

</div>

<div style="float:left;width:40%;">

To obtain **coordinates of a point in polar coordinates,**<br>$\quad$ we need to determine
* $ \quad r      \quad$ the distance of the point from the origin
* $ \quad \theta \quad$ the angle with respect to the x axis
</div>
<div style="float:left;width:40%;padding-left:1cm;">

**Congruent triangles** yield lengths and angles:<br>
$\quad$ Given a line $\left( L \right)$ and a vector $\vec{OA}$,<br>
$\qquad$ construct the mirror image $\tilde{A}$ of $A$ with respect to $\left( L \right)$

The triangles $O A B$ and $O \tilde{A} B$ are congruent, thus:
* $\vert O \tilde{A} \vert \ =\ \vert O {A} \vert$
* $\angle A O B \ =\  \angle \tilde{A} O B$
</div>

## 2.2 Dilation

<div style="float:left;width:40%;padding-right:1cm;">
$$
\begin{align}
    \left( \begin{matrix} 1 \\ 0 \end{matrix} \right) \xrightarrow{D} &  \left( \begin{matrix} \alpha \\ 0 \end{matrix} \right) \\
    \left( \begin{matrix} 0 \\ 1 \end{matrix} \right) \xrightarrow{D} &  \left( \begin{matrix} 0 \\ \beta \end{matrix} \right) \\
\end{align}
$$
</div>
<div style="float:left;width:40%;padding-left:1cm;border-left:2px solid black;height:2.5cm;">
Assembling the dilation matrix, we have
$$
\left( \begin{matrix} x' \\y' \end{matrix} \right) = \left( \begin{matrix}  \alpha & 0 \\ 0 & \beta \end{matrix} \right) \left( \begin{matrix} x\\y\end{matrix} \right)
$$

## 2.3 A rotation in 2D

<div style="float:left;width:40%;padding-right:1cm;">

Consider a rotation $R$ by an angle $\theta$:
$$
\begin{align}
\left( \begin{matrix} 1 \\ 0 \end{matrix} \right) \xrightarrow{R} &  \left( \begin{matrix} \cos \theta \\ \sin \theta \end{matrix} \right)\\
\left( \begin{matrix} 0 \\ 1 \end{matrix} \right) \xrightarrow{R} &  \left( \begin{matrix} \cos ( \theta + \frac{\pi}{2} )\\ \sin ( \theta + \frac{\pi}{2} ) \end{matrix} \right)   = \left( \begin{matrix} - \sin \theta \\ \cos \theta \end{matrix} \right) \\
\end{align}
$$

Assembling the rotation matrix, we have
$$
\left( \begin{matrix} x' \\y' \end{matrix} \right) = \left( \begin{array}{rr}  \cos \theta & - \sin \theta\\ \sin \theta & \cos \theta \end{array} \right) \left( \begin{matrix} x\\y\end{matrix} \right)
$$
</div>
<div style="float:left;width:40%;padding-left:2cm;border-left:2px solid black;height:5.8cm;">
<img src="Figs/RotationMatrix.svg" width="300">
</div>

#### **Example**: $\theta = 30^{\circ},\;\; x\ = 1; y \ = -1$

$$
\left( \begin{matrix} x' \\y' \end{matrix} \right) = \frac{1}{2}\left( \begin{array}{rr}  \sqrt{3} & -1\\ 1 & \sqrt{3} \end{array} \right) \left( \begin{array}{r} 1 \\ -1 \end{array} \right) = \frac{1}{2}\left( \begin{array}{r} 1+\sqrt{3} \\ 1-\sqrt{3}\end{array} \right)
$$

```julia
pr( L"Linear\ Transformation\ \mathbb{R}^2\ \rightarrow\ \mathbb{R}^2:\; rotation\ by\ \theta\ radians", sz=12 )
pr( "Plot on different sets of axes", sz=12 )
println()

vecs = [ 1  0; 0 1]

 = 26
rvecs =  [ cosd() -sind(); sind() cosd()] * vecs

x_vals  = [0 0 ; vecs[1,:]' ]
y_vals  = [0 0 ; vecs[2,:]' ]
rx_vals = [0 0 ; rvecs[1,:]']
ry_vals = [0 0 ; rvecs[2,:]']

plot(
plot(x_vals, y_vals, arrow = true, color = :blue,
     legend = :none, xlims = (-1.5, 1.5), ylims = (-.1, 1.5), title = "Original vectors",
     annotations = [(1.25, 0.2, "[1,0]"),
                    (0.4, 1.2,    "[0,1]")],
     xticks = -5:1:5, yticks = -5:1:5, axis_ratio=:equal,
     framestyle = :origin),
plot(rx_vals, ry_vals, arrow = true, color = :red,
     legend = :none, xlims = (-1.5, 1.5), ylims = (-.1, 1.5), title="Rotated vectors",
     annotations = [(1., 0.65, L"[\cos \theta, \sin \theta]"),
                    (-0.9, 1.1, L"[-\sin \theta, \cos \theta]")],
     xticks = -5:1:5, yticks = -5:1:5, axis_ratio=:equal,
     framestyle = :origin), size=(700,250)
)
```

Observe that the triangles $OPA$ and $OP\tilde{A}$ are congruent:
* The angles $\angle OPA = \angle OP\tilde{A} = 90^\circ$
* The side $PC$ is in common
* The side $PA = P\tilde{A}$ by construction
</div>

                                $P$ has polar coordinates $r= sin \phi,\ \theta = \phi$
</div>
</div>

Consequently,
$\qquad\left.
\begin{align}
\begin{pmatrix} 1\\0\end{pmatrix} &\rightarrow \begin{pmatrix} \quad \cos (2\phi)\\ \quad \sin (2\phi)\end{pmatrix} \\
\begin{pmatrix} 0\\1\end{pmatrix} &\rightarrow \begin{pmatrix} \quad \sin (2\phi)\\ \; -\cos (2\phi)\end{pmatrix} \\
\end{align} \qquad \right\} \quad \Rightarrow \quad\qquad
\begin{pmatrix} x' \\ y' \end{pmatrix}
= \begin{pmatrix} \cos (2\phi) & \sin (2\phi) \\ \sin (2\phi) & -\cos (2\phi) \end{pmatrix}
\begin{pmatrix} x \\ y \end{pmatrix}
$

## 2.5 Orthogonal Projection Onto a Line

The previous figure still applies: The orthogonal projection of $A$ onto the line $(L)$ is $P$.<br>
$\quad$ Using the right triangles $OPA$ and $OP\tilde{A}$ to determine the lengths of $OP$, we have<br><br>
$\qquad\left.
\begin{align}
\begin{pmatrix} 1\\0\end{pmatrix} & \rightarrow \cos \phi \begin{pmatrix} \cos \phi \\ \sin \phi \end{pmatrix}\\
\begin{pmatrix} 0\\1\end{pmatrix} & \rightarrow \sin \phi \begin{pmatrix} \cos \phi \\ \sin \phi \end{pmatrix} \\
\end{align} \qquad \right\} \quad \Rightarrow \quad\qquad
\begin{pmatrix}x' \\ y' \end{pmatrix} =
\begin{pmatrix} \cos^2 \phi & \cos \phi \sin\phi \\ \cos\phi \sin\phi & \sin^2 \phi \end{pmatrix}
\begin{pmatrix}x \\ y \end{pmatrix}
$

## 2.6 Combine Linear Transformations

Consider the transformation $T$ consisting of
* a rotation $R$ by $\theta = 30^\circ$ followed by
* an orthogonal projection $P$ onto the line $\;y = x\ \tan 60^\circ$

----
<div style="float:left;padding-right:1cm;">
The rotation matrix is given by<br><br>

$$R = \left( \begin{array}{rr}   \cos 30^\circ & - \sin 30^\circ \\ \sin 30^\circ & \cos 30^\circ \end{array}\right)
 = \frac{1}{2}\left( \begin{array}{rr}  \sqrt{3} & -1\\ 1 & \sqrt{3} \end{array} \right)
$$
</div>
<div style="float:left;padding-left:1cm;border-left:2px solid black;">
The orthogonal projection matrix is given by<br><br>

$$P =
\begin{pmatrix} \cos^2 60^\circ & \sin 60^\circ \cos 60^\circ \\ \sin 60^\circ \cos 60^\circ & \sin^2 60^\circ \end{pmatrix}
= \frac{1}{4} \begin{pmatrix} 1 & \sqrt{3} \\ \sqrt{3} & 3 \end{pmatrix}
$$
</div>

<div style="width:60%;border:2px solid black;padding:10px 0px;">
$$%\boxed{
    \Large{
    x \xrightarrow{\;\hat{x} = R x\;}
            \hat{x} \xrightarrow{\;\tilde{x} = P \hat{x}\;}
            \tilde{x}
     }
%}
\qquad \text{ thus } \tilde{x} = P \bigg( R x \bigg)
$$
</div>

$\therefore \quad T = P R = \frac{1}{4} \begin{pmatrix} 1 & \sqrt{3} \\ \sqrt{3} & 3 \end{pmatrix}
\frac{1}{2}\left( \begin{array}{rr}  \sqrt{3} & -1\\ 1 & \sqrt{3} \end{array} \right)
= \frac{1}{4} \begin{pmatrix} \sqrt{3} & 1 \\ 3 & \sqrt{3}\end{pmatrix}$

**Remarks:**
* The transformation $R P \ne P R$
* $T$ takes any vector onto the line $y = x \tan 60^\circ$, thus $T$ is **not ONTO.**

# 3. Combining Translations and Linear Transformations

Consider the reflection through the line $\quad x_1 - x_2 = 1$.<br>

Since this line does not include the origin, we need three steps:
* move the origin to lie on the line
* apply the reflection
* move the origin back to its original location

## 3.1 Direct Computation

#### **Translation** $\mathbf{T}$

The translation $T$ defined by $\begin{pmatrix} {x}_1 \\ {x}_2\end{pmatrix} = \begin{pmatrix} \tilde{x}_1 + 1 \\ \tilde{x}_2\end{pmatrix} = \begin{pmatrix} \tilde{x}_1 \\ \tilde{x}_2\end{pmatrix} + \begin{pmatrix} 1 \\ 0 \end{pmatrix}$<br><br>
$\quad$ moves the origin onto the line:
$\quad
x_1 - x_2 = 1 \Leftrightarrow \tilde{x}_1 = \tilde{x}_2.
$<br><br>
The line bisects the first quadrant of the $\tilde{x}_1, \tilde{x}_2$ coordinate system.

#### **Reflection** $\mathbf{R}$

$$
\left.
\begin{align}
\begin{pmatrix} 1\\0\end{pmatrix} & \xrightarrow{T} \begin{pmatrix}0\\1\end{pmatrix} \\
\begin{pmatrix} 0\\1\end{pmatrix} & \xrightarrow{T} \begin{pmatrix}1\\0\end{pmatrix}
\end{align}
\; \right\}
\;\Leftrightarrow \; R = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}
$$

#### **Combined Transformation**

The combined transformation is given by
$\quad x \xrightarrow{\;\; x' \ =\ T x \;\;} x'
\xrightarrow{\;\; x'' \ =\ R x' \;\;} x''
\xrightarrow{\;\; \tilde{x} \ =\ T^{-1} x'' \;\;} \tilde{x} $<br><br>

$\quad \begin{align}
\begin{pmatrix} \tilde{x}_1 \\ \tilde{x}_2 \end{pmatrix} = &\ 
R \;
\bigg( \begin{pmatrix} {x}_1 \\ {x}_2 \end{pmatrix} - \begin{pmatrix} 1 \\ 0 \end{pmatrix} \bigg)
+ \begin{pmatrix} 1 \\ 0 \end{pmatrix} \\
= & \ R \begin{pmatrix} {x}_1 \\ {x}_2 \end{pmatrix} - R \begin{pmatrix} 1 \\ 0 \end{pmatrix} + \begin{pmatrix} 1 \\ 0 \end{pmatrix} \\
= & \ \begin{pmatrix} {x}_2 \\ {x}_1 \end{pmatrix} + \begin{pmatrix} 1 \\ -1 \end{pmatrix}
\end{align}
$

## 3.2 Homogeneous Coordinates

With homogeneous coordinates, the three transformations can be written with matrices:<br><br>
$\qquad T = \left( \begin{array}{cc|c} 1 & 0 & 1 \\ 0 & 1 & 0 \\ \hline 0 & 0 & 1 \end{array} \right),
 \quad R = \left( \begin{array}{cc|c} 0 & 1 & 0 \\ 1 & 0 & 0 \\ \hline 0 & 0 & 1 \end{array} \right)
$

The combined transformation is thus given by<br><br>
$\quad \begin{align}
\begin{pmatrix} \tilde{x}_1 \\ \tilde{x}_2 \\ \hline 1 \end{pmatrix} \;
= & \; T^{-1} R T\ \begin{pmatrix} x_1 \\ x_2 \\ \hline 1 \end{pmatrix} \\[2mm]
= & \; \left( \begin{array}{rr|r} 1 & 0 & -1 \\ 0 & 1 & 0 \\ \hline 0 & 0 & 1 \end{array} \right)
       \left( \begin{array}{cc|c} 0 & 1 &  0 \\ 1 & 0 & 0 \\ \hline 0 & 0 & 1 \end{array}  \right)
       \left( \begin{array}{rr|r} 1 & 0 &  1 \\ 0 & 1 & 0 \\ \hline 0 & 0 & 1 \end{array} \right)
\begin{pmatrix} x_1 \\ x_2 \\ \hline 1 \end{pmatrix} \\[2mm]
=& \; \left( \begin{array}{rr|r} 0 & 1 & -1 \\ 1 & 0 &  1 \\ \hline 0 & 0 & 1 \end{array} \right)
\begin{pmatrix} x_1 \\ x_2 \\ \hline 1 \end{pmatrix}
\end{align},$<br><br>
$\quad$ the same result as before.

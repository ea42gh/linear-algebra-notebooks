## 2. Ridge Regression as GEP
Express ridge via eigen-decomposition of $A^T A$.

```python
np.random.seed(0)
A = np.random.randn(100, 5)
x_true = np.arange(5)
b = A @ x_true + 0.5 * np.random.randn(100)

 = 2.0
M = A.T @ A
2, V = eig(M)
2 = np.real(2)
c = V.T @ (A.T @ b)
x_ridge = V @ (c / (2 + ))

print('Ridge solution:', np.round(x_ridge,3))
print('True x:', x_true)
```

## 3. Fisher Discriminant Analysis
Solve between-class vs within-class scatter GEP.

```python
np.random.seed(42)
X0 = np.random.randn(50,3)
X1 = np.random.randn(50,3)+2
mean0, mean1 = X0.mean(0), X1.mean(0)
Sb = np.outer(mean1-mean0, mean1-mean0)
Sw = (X0-mean0).T@(X0-mean0) + (X1-mean1).T@(X1-mean1)
f, wf = eig(Sb, Sw)
w = np.real(wf[:, np.argmax(np.real(f))])
p0, p1 = X0@w, X1@w
plt.hist(p0, alpha=0.6, label='class 0')
plt.hist(p1, alpha=0.6, label='class 1')
plt.legend(); plt.title('Fisher separation'); plt.show()
```

## 4. Interactive Parameter Study for Ridge 
Adjust ridge regularization interactively.

```python
import pandas as pd

class RidgeExplorer(pn.viewable.Viewer):
    def __init__(self, A, b, **params):
        super().__init__(**params)
        self.A, self.b = A, b
        self.slider = pn.widgets.FloatSlider(name='', start=0.01, end=10.0, value=1.0)
        self.panel = pn.bind(self._view, self.slider)

    def _view(self, ):
        A, b = self.A, self.b
        2, V = eig(A.T @ A)
        2 = np.real(2)
        V = np.real(V)
        c = V.T @ (A.T @ b)
        xr = V @ (c / (2 + ))
        return pn.Column(
            pn.pane.Markdown(f"**** = {:.2f}"),
            pn.pane.DataFrame(pd.DataFrame({'Coef': xr}), width=250)
        )

    def __panel__(self):
        return pn.Column(self.slider, self.panel)
```

```python
import numpy as np

# Random seed for reproducibility
np.random.seed(42)

# Design matrix A: 100 samples, 5 features
A = np.random.randn(100, 5)

# True coefficients
x_true = np.array([1, 2, 3, 4, 5])

# Response vector with noise
b = A @ x_true + 0.5 * np.random.randn(100)


ridge_app = RidgeExplorer(A, b)
ridge_app.servable()
```

## 1.2 Anticommutativity

Since $a \wedge b$ and $b \wedge a$ have opposite orientation, set
$\quad$ $\boxed{ a \wedge b \;=\; - b \wedge a}$

##### **Special Case** $\mathbf{a = b}$

Since $\; a \wedge a = - a \wedge a \;\;\qquad\qquad\qquad\qquad$ set $\quad \boxed{a \wedge a = 0}$

## 1.4 Generalization to More Vectors

### 1.4.1 Blades

<div style="float:left;background-color:#F2F5A9;color:black;padding-right:0.5cm;width:45%;height:2cm;">

**Definition:** Let $a_1, a_2, \dots a_k$ be vectors in $\mathbb{R}^N$.<br>
$\quad$ The product $a_1 \wedge a_2 \dots \wedge a_k$ is a **blade** or a **k-blade**.
</div>
<div style="float:left;padding-left:0.5cm;width:50%;">

We generalize the wedge product to more vectors,<br>
$\quad$ using similar geometric arguments<br>
$\quad$ to produce a product computing **hypervolumes**
</div>

The wedge product is **associative** and **linear**.

$\quad \boxed{ 
\begin{align} ( & a_1 \wedge a_2 ) \wedge a_3 & =&\; a_1 \wedge ( a_2 \wedge a_3 ) \\
& (\alpha a_1) \wedge a_2 \wedge a_3 & =& \; \alpha\;  a_1 \wedge a_2 \wedge a_3 \\
& (b_1 + c_1) \wedge a_2 \wedge a_3 & =&\; b_1 \wedge a_2 \wedge a_3\ +\ c_1 \wedge a_2 \wedge a_3
\end{align} }
$

etc...

### 1.4.2 Interchanging Vectors, Repeated Vectors

#### **Moving Vectors in a Blade**

<div style="float:left;width:45%;">

> **Example: Move $a_1$ to the right**
>
> $\quad$ $\begin{align}
\xi \; = & \; \color{red}+ \color{red}{a_1} \wedge a_2 \wedge a_3 \wedge a_4 \\
 \; = & \;   \color{red} - a_2 \wedge \color{red}{a_1} \wedge a_3 \wedge a_4 \\
 \; = & \;  \color{red} + a_2 \wedge a_3 \wedge \color{red}{ a_1} \wedge a_4 \\
 \; = & \;  \color{red} - a_2 \wedge a_3  \wedge a_4 \wedge \color{red}{ a_1}
\end{align}
$
</div>
<div style="float:left;width:45%;">

> **Example: Interchange $a_1$ and $a_4$**
>
> $\quad$ $\begin{align}
\zeta \; = & \; \color{red}+{a_1} \wedge a_2 \wedge a_3 \wedge \color{red}{a_4}  \wedge a_5 \\
 \; = & \; \color{red}+ \color{red}{a_1} \wedge \color{red}{a_4} \wedge a_2 \wedge a_3 \wedge a_5  \\
 \; = & \; \color{red}- \color{red}{a_4} \wedge \color{red}{a_1} \wedge a_2 \wedge a_3 \wedge a_5 \\
 \; = & \; \color{red}- {a_4} \wedge a_2 \wedge a_3  \wedge \color{red}{ a_1} \wedge a_5
\end{align}
$
</div>

#### **Repeated Vectors in a Blade**

<div style="float:left;width:45%;">

> **Example: $a_2$ appears more than once**
>
> $\quad$ $\begin{align}
\xi \; = & \; a_1 \wedge \color{red}{a_2} \wedge a_3 \wedge a_4\wedge \color{red}{a_2}\\
 \; = & \;  a_1 \wedge(  \color{red}{a_2} \wedge \color{red}{a_2} ) \wedge a_3 \wedge a_4 \\
 \; = & \;  0
\end{align}
$<br>
> $ $
</div>
<div style="float:left;width:45%;">

> **Example: $b$ appears more than once**
>
> $\quad$ $\begin{align}
\zeta \; =  \; & a_1 \wedge \color{red}{( a_2+b )} \wedge a_3 \wedge a_4 \wedge \color{red}{b}\\
 \; =  \; & a_1 \wedge \color{red}{a_2}\wedge a_3 \wedge a_4 \wedge \color{red}{b} \\
        + & a_1 \wedge \color{red}{b} \wedge a_3 \wedge a_4 \wedge \color{red}{b}
 \\
  \; =  \;&  a_1 \wedge \color{red}{a_2} \wedge a_3 \wedge a_4 \wedge \color{red}{b}
\end{align}
$
</div>

## 1.5 Example Computations

We can do algebraic computations similar to scalar algebra, with the proviso that
* the order of the vectors must be maintained: **interchanging vectors results in a sign change**
* **repeated vectors simplify to zero**

### 1.5.1 Examples with 2 Vectors

#### **Example 1**

<div style="float:left;width:50%;">
Let $a = 4 e_1 + 2 e_2,  b = 3 e_1 - e_2$

$\quad \begin{align}
a \wedge b =&\  (4\ e_1 + 2\ e_2 ) \wedge ( 3\ e_1 - e_2 ) \\
           =&\ 4\ e_1 \wedge ( 3\ e_1 - e_2 ) + 2\ e_2 \wedge ( 3\ e_1 - e_2 ) \\
           =&\ 12\ e_1 \wedge e_1 -4\ e_1 \wedge e_2 + 6\ e_2 \wedge e_1 - 2\ e_2 \wedge e_2 ) \\
           =&\ -4\ e_1 \wedge e_2 -6\ e_1 \wedge e_2 \\
           =&\ -10\ e_1 \wedge e_2
\end{align}$

**Notation:** This will be easier to read<br>
$\qquad$ if we **introduce subscripts**, e.g. $\;\;\mathbf{e_{2 1} = e_2 \wedge e_1}$<br>
$\qquad$ Thus, $\; a \wedge b = -10\ e_{1 2}$

</div>
<div style="float:left;width:40%;padding-left:1cm;">

**Interpretation:**
* The area defined by $\mathbf{a \wedge b}$ is 10 times larger<br> than the area of $\mathbf{e_1 \wedge e_2}$
* The areas have opposite orientation

#### **Example 2**

<div style="float:left;width:35%;padding-right:0.5cm;">
Let $a = 4 e_1 + 2 e_2 + e_3, \;\; b = 3 e_1 - e_3$

$\quad \begin{align}
a \wedge b =&\  (4\ e_1 + 2\ e_2 + e_3 ) \wedge ( 3\ e_1 - e_3 ) \\
           =&\ -4\ e_{1 3} + \left( 6 e_{2 1} - 2 e_{2 3} \right) + 3 e_{3 1} \\
           =&\ -7\ e_{1 3} -6 e_{1 2} -2 e_{2 3}
\end{align}$

**Remark:**<br>
$\qquad a \wedge ( b + c ) \; = \; a \wedge b + a \wedge c$<br>
$\quad$ does not require that $a, b, c$ be coplanar:<br>
$\quad$ the area has been decomposed into<br>
$\qquad$ **areas in the coordinate planes.**
</div>
<div style="float:left;width:50%;padding-left:1cm;border-left: 2px solid black;">

**The cross product obeys the exact same rules**<br><br>
$\qquad\quad a \times b = -7\ e_1 \times e_3\ - 6\ e_1 \times e_2 -2\ e_2 \times e_3$

$\qquad$ It does hoever introduce **the vector orthogonal to a plane:**<br>
$\qquad\qquad$ $e_1 \times e_2 = e_3, \; e_2 \times e_3 =  e_1, \; e_3 \times e_1 = e_2$

$\qquad$ Thus, $\;\; a \times b = 7 e_2 -6 e_3 -2 e_1 \;=\; -2 e_1 + 7 e_2 - 6 e_3$
    
<strong style="color:red;">$\qquad$ The cross-product is only defined in 3D!</strong>
</div>

### 1.5.2 Example with 3 Vectors

<div style="float:left;width:35%;padding-right:1.5cm;">
Let $\quad \left\{ \begin{align}
a &= 4 e_1 + 2 e_2 + e_3\\
b &= 3 e_1 - e_2 + e_3\\
c &= e_1 + e_2 + 2 e_3
\end{align} \right.$
<br><br>
$\begin{align}
a \wedge b \wedge c \
= & \ \;\;   \left( 4 e_1 + 2 e_2 + \;e_3 \right) \\
&            \wedge \left( 3 e_1 -\; e_2 +\; e_3   \right) \\
&            \wedge \left(\; e_1 +\; e_2 + 2 e_3   \right) \\
= & \ \;\;\; 4 e_1 \wedge \left(  - e_2 + e_3   \right) \wedge \left(  e_2 + 2 e_3 \right) \\
&     + 2 e_2  \wedge \left(  3 e_1  + e_3   \right) \wedge \left( e_1 + 2 e_3 \right) \\
&     + \; e_3 \wedge \left(  3 e_1  - e_2   \right) \wedge \left( e_1 + e_2 \right) \\
= & \ \;\;\; 4 e_1 \wedge (-2 -1)\ e_{2 3} \\
& + 2 e_2 \wedge (6-1)\ e_{1 3} \\
&+ \; e_3 (3 + 1)\ e_{1 2} \\
= &\ \; -12\ e_{1 2 3} - 10\ e_{1 2 3} + 4\ e_{1 2 3} \\
= &\ \;  -18 e_{1 2 3}
\end{align}
$
</div>
<div style="float:left;height:9cm;width:50%;padding-left:1cm;border-left: 2px solid black;">

**Remarks:**
* once we pick **an entry** in $a$, it **multiplies a wedge product of two vectors**
* we order the $e_i$ in the blade in **numerical order**
* $k$ vectors expanded into $k$ entries results in
    <strong style="color:red;">a single term $\; e_{1 2 \dots k}$</strong>
<br><br><br>
On the right, we have used the expansion<br><br>
$\quad$
$\begin{align}
\zeta & = \;\;\; \left( \alpha e_i + \beta e_j \right) \\
      & \;\;\;  \wedge \left( \gamma e_i + \delta e_j \right) \\
      & = \;\;\; \alpha \delta\ e_i \wedge e_j  + \beta \gamma \ e_j \wedge e_i \\
      & = \;\;\; \left( \alpha \delta - \beta \gamma \right) \ e_{ i j }
\end{align}$
</div>

## 1.6 Hyper-volumes: the Determinant

<div style="float:left;background-color:#F2F5A9;color:black;padding-right:0.3cm;">

**Definition:** Let $a_1, a_2, \dots a_k$ be vectors in $\mathbb{R}^N$.<br>
$\quad$ with each of the vectors $a_i$ given as a linear combination <br>$\quad$ of the standard basis $e_1, e_2, \dots e_k$ in $\mathbb{R}^N$.<br>
$\quad$ Then $\; a_1 \wedge a_2 \dots \wedge a_k = \Delta\ \color{red}{e_{1 2 \dots k }}$<br><br>
$\quad$ The scalar $\Delta$ is the **determinant** of the vectors $a_i, i = 1,2, \dots k$
</div>
<div style="float:left;width:35%;padding-left:0.5cm;">
    
**Remark:**<br>
 $\quad$ the determinant is the **hyper-volume**<br>$\quad$ relative to the **unit hypercube**
<div>

# 2. The Determinant Via Laplace Expansion

## 2.1 Using the first Row

$\begin{align}
& \left| \begin{matrix}a \end{matrix} \right|                       &=&\;   a \\
& \left| \begin{matrix}a_1 & a_2 \\ b_1 & b_2 \end{matrix} \right|  &=&\;\;
         a_1 \left| \begin{matrix} b_2 \end{matrix} \right| - a_2  \left| \begin{matrix} b_1 \end{matrix} \right|  \\
& \left| \begin{matrix}a_1 & a_2 & a_3 \\ b_1 & b_2 & b_3 \\ c_1 & c_2 & c_3 \end{matrix} \right|  &=&\;\;
         a_1 \left| \begin{matrix} b_2 & b_3 \\ c_2 & c_3 \end{matrix} \right|
       - a_2  \left| \begin{matrix} b_1 & b_3 \\ c_1 & c_3 \end{matrix} \right|
       + a_3  \left| \begin{matrix} b_1 & b_2 \\ c_1 & c_2 \end{matrix} \right|  \\
& \left| \begin{matrix}a_1 & a_2 & a_3 & a_4 \\ b_1 & b_2 & b_3 & b_4 \\ c_1 & c_2 & c_3 & c_4 \\ d_1 & d_2 & d_3 & d_4\end{matrix} \right|  &=&\;\;
         a_1 \left| \begin{matrix} b_2 & b_3 & b_4 \\ c_2 & c_3 & c_4 \\ d_2 & d_3 & d_4 \end{matrix} \right|
       - a_2 \left| \begin{matrix} b_1 & b_3 & b_4 \\ c_1 & c_3 & c_4 \\ d_1 & d_3 & d_4 \end{matrix} \right|
       + a_3 \left| \begin{matrix} b_1 & b_2 & b_4 \\ c_1 & c_2 & c_4 \\ d_1 & d_2 & d_4 \end{matrix} \right|
       - a_4 \left| \begin{matrix} b_1 & b_2 & b_3 \\ c_1 & c_2 & c_3 \\ d_1 & d_2 & d_3 \end{matrix} \right| \\
& \dots &
\end{align}$

#### **3x3 Example**

$\qquad\begin{align}
\left| \begin{matrix} \color{red}1 & \color{red}2 & \color{red}4 \\ 0 & 1 & 3 \\ 2 & -1 & 0 \end{matrix} \right| &=
    \color{red}1^ \ \left| \begin{matrix} 1 & 3 \\ -1 &  0 \end{matrix} \right|
   -\color{red}2 \ \left| \begin{matrix} 0 & 3 \\  2 &  0 \end{matrix} \right|
   +\color{red}4 \ \left| \begin{matrix} 0 & 1 \\  2 & -1 \end{matrix} \right| \\
   &= \color{red}1 (0+3) - \color{red} 2 ( 0-6) + \color{red}4 (0-2) \\
   &= 7
\end{align}$

## 2.2 Using Different Rows or Columns

#### **Different Rows**

We know that moving a vector in the wedge product $a_1 \wedge a_2 \dots \wedge a_N$ changes the sign.<br>
$\quad$ This corresponds to **moving a row in the matrix.**<br>
$\quad$ Rather than physically moving the row, we just have to **account for the sign change**<br>



The pattern of signs is a **checkerboard pattern:**

$\qquad
\left( (-1)^{i+j} \right) = \begin{pmatrix} +1 & -1 & +1 & \dots \\
                                           -1 & +1 & -1 & \dots \\
                                           +1 & -1 & +1 & \dots \\
                                           \dots & \dots & \dots & \dots \end{pmatrix}
$

#### **Different Columns**

The systematic expansion of the wedge product allows us to **expand by columns instead.**<br>
$\quad$ The **sign pattern stays the same.**

## 2.3 Some Easy Determinants

#### **Row (or Column) of Zeros**

> **Example**
>
> $\qquad\left| \begin{matrix} 1 & 2 & \color{red}0 & 3 \\
                          4 & 2 & \color{red}0 & 5 \\
                          8 & 2 & \color{red}0 & 7 \\
                          4 & 1 & \color{red}0 & 6 \end{matrix} \right| = 0$

* The determinant of a matrix with a row or column of zeros is **zero.**

#### **A Row/Column Is a Linear Combination of other Rows/Columns**

Transcribing $a_1 \wedge \dots a_{i-1} \wedge \left( \sum_{j\ne i} \alpha_j a_j \right) \wedge a_{i+1} \dots \wedge a_k = 0,$<br>
$\quad$ to matrix form, we see that such a determinant is zero.

> **Example:**
>
> $\qquad\left| \begin{matrix}  1 & 2 & \color{red}3 & 3 \\
                          4 & 2 & \color{red}6 & 5 \\
                          8 & 2 & \color{red}{10} & 7 \\
                          4 & 1 & \color{red}5 & 6 \end{matrix} \right| = 0,$
> $\quad$ since columns 1 and 2 add up to the third column.

* The determinant of a matrix such that some row(column) is a linear combination<br>$\quad$ of other rows(columns) is **zero.**

#### **Diagonal Matrices**

> **Example:**<br>
> $\qquad\left| \begin{matrix} \color{red}3 & 0 & 0 & 0 \\ \color{red}0 & 5 & 0 & 0 \\ \color{red}0 & 0 & 2 & 0 \\ \color{red}0 & 0 & 0 & -1 \end{matrix} \right| = 3\times 5\times 2\times(-1) = -30$
>
> as can be easily seen by expanding using the Laplace expansion<br>
> $\quad$ with the first row in each of the determinants in turn.

* In general, the **determinant of a diagonal matrix** is the **product of the diagonal elements**

#### **Triangular Matrices**

The previous example generalizes to triangular matrices:

> **Example:**
>
>$\qquad\left| \begin{matrix} \color{red}3 & \color{red}0 & \color{red}0 & \color{red}0 \\ 4 & 5 & 0 & 0 \\2 & 1 & 2 & 0 \\ 8 & -9 & 0 & -1 \end{matrix} \right| = 3\times 5\times 2\times(-1) = -30$

where we expand using the first row in each successive matrix.

In general, the **determinant of a triangular matrix** is the **product of the diagonal elements**

#### **Row Exchange Matrices**

> **Example:**<br>
>
> The the following permutation matrix would interchange rows 2 and 4<br>
> $\quad$ whem multiplying some matrix from the left.
>
> $\qquad\left| \begin{matrix} 1 & 0 & 0 & 0 \\
                          0 & \color{red}0 & 0 & \color{red}1 \\
                          0 & 0 & 1 & 0 \\
                          0 & \color{red}1 & 0 & \color{red}0 \end{matrix} \right| =
                          \left| \begin{matrix} 0 & 1 \\ 1 & 0 \end{matrix} \right| = -1
$
>
> as can be easily seen by expanding using the Laplace expansion<br>
> $\quad$ using the rows or columns with 1 on the diagonal.


* A matrix that exchanges a **single pair of rows** has determinant $\mathbf{-1}$
* In general, a permutation matrix has determinant +1 or -1.

## 2.5 Bad News: Computation is Not Feasible

The number of multiplies for a matrix of size $N \times N$ is $N!$

$\quad$ For $N=\;5$, this is $\qquad\quad\qquad$ 120 $\;\;=\;\; 1.2\ 10^{2}\;\quad$  (hundreds)<br>
$\quad$ For $N=10$, this is $\qquad\quad$ 3,628,800 $\;\;\approx\;\; 3.6\ 10^{6}\;\quad$  (millions)<br>
$\quad$ For $N=15$, this is $\;$ 1,307,674,368,000  $\;\;\approx\;\; 1.3\ 10^{12}\quad$  (trillion)

$\quad$ **We will need a better algorithm.**

# 3. Take Away

* The **wedge product** captures the notion of a **hyper-volume** associated with a set of vectors
    * a **k-blade** can be thought of as an oriented **hyperplane**
    * in 3D, the wedge product is closely related to the **cross-product**

* The determinant is the hypervolume of a wedge product with respect to a reference hypervolume
    * transcription of the wedge product to computations with matrix entries<br>
    $\quad$ resulted in the **Laplace Expansion**

* Matrices with easily computed determinants include<br>
    $\quad$ **each of the elementary step matrices** from Gaussian Elimination:<br><br>
    * **Elimination Matrices** are unit lower triangular and have determinant 1
    * **Row Exchange matrices** have determinant -1
    * **Scaling Matrices** are diagonal and have determinant equal to the product of the scale factors

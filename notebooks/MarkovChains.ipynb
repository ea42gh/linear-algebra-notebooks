## 1.1 Create Random 2x2 Matrices with Given Eigenvalues

```python
def random_unit_lower_triangular(n):
    L = np.tril(np.random.rand(n, n))
    np.fill_diagonal(L, 1)
    return L
# -------------------------------------------------------------------------------------------------
def matrix_A( e1, e2 ):
    """random 2x2 matrix with eigenvalues e1, e2"""
    S = random_unit_lower_triangular(2) @ random_unit_lower_triangular(2).T
    Sinv = np.linalg.inv(S)
    Lambda = np.diag([e1,e2])
    return S @ Lambda @ Sinv
# -------------------------------------------------------------------------------------------------
def rotation_matrix_A( theta ):
    # eigvals  e^(i theta)  e^(-i theta)
    # eigvecs  [i, 1]       [-i 1]
    c=np.cos(theta)
    s=np.sin(theta)
    return np.array([[c,-s],[s, c]])
```

## 1.2 PhasePortrait

```python
class PhasePortrait:
    def __init__(self, A):
        self.set(A)

    def set(self, A):
        self.N                 = A.shape[1]
        self.A                 = A
        self.evals, self.evecs = np.linalg.eig(A)

    def iterates(self, v, N):
        v = np.array(v)
        v = v.copy()
        x = [v[0]]
        y = [v[1]]
        for _ in range(N):
            v = self.A @ v
            v = v.copy()
            x.append(v[0]); y.append(v[1])
        return x,y

    def dydx(self,x,y, epsilon=1e-6, max_slope=1e6 ):
        x, y           = np.asarray(x), np.asarray(y)
        original_shape = x.shape
        points         = np.vstack((x.ravel(), y.ravel()))
        next_points    = self.A @ points
        moved_by       = next_points - points
        dx             = moved_by[0,:]
        dy             = moved_by[1,:]

        if x.ndim == 0:     return dx[0], dy[0]                 # Scalar input
        elif x.ndim == 1:   return dx, dy                       # Vector input
        else:               return dx.reshape(original_shape),\
                                   dy.reshape(original_shape)   # Meshgrid or higher dimensional input
    # ========================================================================================
    @staticmethod
    def background(func, bounds=(-5,5), size=(500,500)):
        """
        Given the function dx,dy = f(x,y) that returns the gradient of f(x,y) at point (x,y)

           bg,vec,xaxis_line,yaxis_line = background()

        returns a vector field representation of the slopes, and
                a set of axis lines for x and y within specified bounds
        """

        # compute the data
        vals = np.linspace( *bounds, num=150 )
        X, Y = np.meshgrid(vals, vals)

        clines  = func(X,Y)                           # f(x,y)
        theta   = np.arctan2( clines[1], clines[0])   # angle of the slope y' at x,y
        # Obtain the vector field (subsample the grid)
        h,w=size
        vf_opts = dict(height=h, width=w, xticks=9, yticks=9, alpha=0.3, muted_alpha=0.05)
        vec_field = hv.VectorField((vals[::3],vals[::3], theta[::3,::3],0*clines[0][::3,::3]+1),
                                   label='vector_field' ).options(**vf_opts)

        # finally, we add the axes as VLine, HLine and return an array of the plot Elements
        hv_opts = dict( color='k', alpha=0.8, line_width=1.5)
        return [vec_field, hv.HLine(0).options(**hv_opts),
                    hv.VLine(0).options(**hv_opts)]

    def dbg(self, a):
        print(  self.re_indices, self.cx_indices )
        print("a =")
        for i in range(self.N): print(f".  ", a[i])
        print( "evals =")
        for i in range(self.N): print(f".  ", self.evals[i])
        print( "evecs =")
        for i in range(self.N): print(f".  ", self.evecs[:,i])

# Usage:
#phase_portrait = PhasePortrait( np.array([[0,-1,0],[1,0,0],[0,0,2]]) )
#phase_portrait.phase_curve( (1,2,3) )  # computr a phase curve starting at (1,2,3)
#phase_portrait.df_dt( (1,2,3) )  # compute the t derivative at 0
```

```python
def show_phase_portrait(A=np.array([[0,-1],[1,0]]), points=[(-2,-2)], N_iterates=10 ):
    phase_portrait_2D = PhasePortrait( A )
    title = f"Eigenvalues {phase_portrait_2D.evals[0]:.3}, {phase_portrait_2D.evals[1]:.3}"

    hl = PhasePortrait.background(phase_portrait_2D.dydx,bounds=(-3,3),)

    for i,v in enumerate(points):
        x,y = phase_portrait_2D.iterates(v, N_iterates-1)
        df = pd.DataFrame({ 'x': x, 'y': y, 'index': np.arange(N_iterates) })
        hl = hl +\
         [ hv.Curve(df, kdims=['x'], vdims=['y'], label=str(i))\
             .opts(line_width=1.5, color='blue', muted_alpha=0, title = title),
           hv.Scatter(df, kdims=['x'], vdims=['y', 'index'], label=str(i))\
             .opts( color='index', cmap='plasma', clim=(0, N_iterates-1), muted_alpha=0, size=3, tools=['hover'], colorbar=True,),
         ]

    display(hv.Overlay(hl).opts(show_legend=True, legend_position='left', width=600, height=400))
    return df
```

## 2.1 Definitions

<div style="background-color:#F2F5A9;color:black;height:1cm;">

**Definition:**
A probability vector $v$ is a vector whose elements are nonnegative real numbers that sum to 1.
</div>

<div style="background-color:#F2F5A9;color:black;">

**Definition:**
* A **right stochastic matrix** is a square matrix of nonnegative real numbers, with each row summing to 1.
* A **left stochastic matrix** is a square matrix of nonnegative real numbers, with each column summing to 1.
* A **doubly stochastic matrix** is a square matrix of nonnegative real numbers with each row and column summing to 1.
</div>

We will use the convention that the columns are probability vectors, i.e. that the stochastic matrices are left stochastic matrices.

**Example:**
$\;\;P = \begin{pmatrix}
.1 & 0 & 0 & .2  \\
.3 & 1 & .9 & .4 \\
.2 & 0 & 0 &  .1 \\
.4 & 0 & .1 & .3
\end{pmatrix}$
 is a stochastic matrix.

**Remark:** By definition the vector of all ones is an eigenvector of a right stochastic matrix with eigenvalue $\lambda = 1$.

<div style="background-color:#F2F5A9;color:black;">

**Definition:** A **Markov chain** is a stochastic process describing a sequence of possible events<br>
$\qquad$ where the probability of each event depends only on the state attained in the previous event.

A **markov chain** It can be represented by:
* A state space $S$
* A transition probability matrix $P$ (a stochastic matrix)
* An initial probability distribution (a probability vector)
</div>

#### Markov Chain Example

Consider the following two examples of the probability of moving between socioeconomic classes.<br>
* There are three possible states: {Rich, Middle Class, Poor}
* The directed edges indicate the probability of transitioning from a given state to another state (not necessarily distinct!)
* For each state, the transition probabilities to another state (including the current state) sum to 1.

<br>The transition probability matrix $P = (p_{i j})$, where index $i$ is an ending state and $j$ is a starting state,<br>
with increasing indices $i$ and $j$ corresponding to each of the states {Rich, Middle Class, Poor} in turn.<br>
I.e., a row of the matrix records the probabilities of incoming transitions.

The first example has
$\;\;P_1 = \begin{pmatrix} 0.8 & 0.2 & 0.0 \\ 0.1 & 0.4 & 0.1 \\ 0.1 & 0.4 & 0.9 \end{pmatrix},\;\;$

while the second more pessimistic example has
$\;\;P_2 =\begin{pmatrix} 0.9 & 0.3 & 0.0 \\ 0.1 & 0.4 & 0.0 \\ 0.0 & 0.3 & 1.0 \end{pmatrix}.$

The matrices $P_1$ and $P_2$ **are (left) stochastic matrices** and express the transition probabilities from state to state over a given period of time.

Note that in the second model $P_2$, the poor remain poor forever!

<div style="background-color:#F2F5A9;color:black;height:1.0cm;">

**Definition:**
A state $i$ is **absorbing** if it is impossible to leave this state.
</div>

Therefore, the state $j$ is absorbing if $P_{j j} = 1$ and $P_{i j} = 0$ for $j \ne i$.<br>
If every state can reach an absorbing state, then the Markov chain is an absorbing Markov chain.

For $P_2$, "poor" is and absorbing state, and the corresponding Markov Chain is an absorbing chain.

We define a **probability vector** to be the state distribution, e.g., $v = \begin{pmatrix} 0.1 \\ 0.5 \\ 0.4 \end{pmatrix}$<br>
expresses that $10$% of a given population is rich, $50$% is middle class, and the remaining $40$% of the population is poor.

The product of a a stochastic matrix with a probability vector yields another probability vector,<br>
representing the state distribution after one transition in the Markov chain:<br>
a given percentage of rich people loose money and join the middle class or become poor, etc.

In our example $\tilde{v} = P_1 v = \begin{pmatrix} 0.17 \\ 0.38 \\ 0.41\end{pmatrix},$ i.e., a new distribution of wealth.

**Remark:** the convention of using left stochasic matrices, i.e., $P_1^t$,<br>
would multiply the matrices by a row vector from the left: $\;\;\tilde{v}^t = v^t P_1$

```python
P_1 = np.array([[0.8,0.2,0.0],[0.1,0.4,0.1], [0.1,0.4,0.9]])
P_2 = np.array([[0.9, 0.3, 0.0],[0.1,0.4,0.0],[0.0,0.3,1.0]])
v   = np.array([0.1,0.5,0.4])
np.array([P_1 @ v, P_2 @ v])
```

Thus the transition from a current probability vector $v_0$ to a new probability vector $v_1 = P v_0$ is a state transition in a Markov chain:<br>
$\qquad v_1 = P v_0$.

The markov chain assumption keeps the transition matrix $P$ constant: we can apply it repeatedly to observe the evolution of a probability vector:<br>
$\qquad v_n = P v_{n-1}, n=1, 2, \dots$ for a given initial probability $v_0$.

(See [**DifferenceEquations.ipynb**](DifferenceEquations.ipynb))

## 2.2 The Evolution of the Probability Vector

We can observe the evolution of the probability vector for a given Markov chain.<br>
Continuing our previous examples
with the stocastic matrices $P_1$ and $P_2$ we see the following:

```python
def chain(P, v, n, state_names):
    results = []
    results.append(v.copy())
    for _ in range(n):
        v = P @ v
        results.append(v.copy())
    return pd.DataFrame( np.array(results), columns=state_names)

df_1 = chain(P_1, v, 100, ['rich', 'middle class', 'poor'])
df_2 = chain(P_2, v, 100, ['rich', 'middle class', 'poor'])
```

```python
h_1 =\
hv.Curve( df_1, 'index', 'rich',         label='rich').opts(title='Transition Matrix P_1', show_grid=True) *\
hv.Curve( df_1, 'index', 'middle class', label='middle class')*\
hv.Curve( df_1, 'index', 'poor',         label='poor')
h_2 =\
hv.Curve( df_2, 'index', 'rich',         label='rich').opts( title='Transition Matrix P_2', show_grid=True)*\
hv.Curve( df_2, 'index', 'middle class', label='middle class')*\
hv.Curve( df_2, 'index', 'poor',         label='poor')

pn.Row((h_1 + h_2).opts("Overlay", legend_position='bottom'))
```

We see that for $P_2$, the absorbing state "poor" eventually accumulates all the probability: $P_2$ has an eigenpair $(1,\begin{pmatrix}0\\0\\1\end{pmatrix})$.

A probability vector that is an eigenvector of a transition matrix with eigenvalue 1 is called a **stationary distribution.**<br>
$\qquad$ It can be shown that all stochastic matrices have a stationary distribution:<br>
$\qquad$ Since $\lambda=1$ is always an eigenvalue, the corresponding Eigenspace basis must have at least one eigenvector.<br>
$\qquad$ The Perron-Frobenius Theorem ensures that one can always choose this basis vector to be a probability vector.

## 2.3 Phase Curves and Phase Portraits

Given a matrix $A$ and a starting point $x_0$, we can compute successive iterates $x_n = A x_{n-1}$.
The resulting sequence of iterates is a **phase plot** for this particular starting point. We can readily make this into a **phase curve** by joining $x_{n-1}, x_n$ with a line segment, illustrating a trajectory.

We can further add arrows indicating the direction of travel of a point on a trajectory, creating a vector field<br>
known as a **phase diagram** or **phase portrait**.

```python
pp = PhasePortrait( np.sqrt(2)/2*np.array([[1,-1],[1,1]] ))

def plot_xy(x,y):
    df = pd.DataFrame({ 'x': x, 'y': y, 'index': np.arange(len(x)) })
    return [ hv.Curve(df, kdims=['x'], vdims=['y'])\
         .opts(line_width=1.5, color='blue', muted_alpha=0),
       hv.Scatter(df, kdims=['x'], vdims=['y', 'index'])\
         .opts( color='index', cmap='plasma', clim=(0, len(x)-1), muted_alpha=0, size=3, tools=['hover'], colorbar=False,),
     ]

def re_portrait(e1,e2,v1,v2,N_iterates):
    bounds = (-3,3)
    pp.set( np.array([[e1,1],[0,e2]]))

    h = PhasePortrait.background(pp.dydx,bounds=bounds) +\
        plot_xy( *pp.iterates((v1,v2), N_iterates-1) )

    return hv.Overlay(h).opts(show_legend=True, legend_position='top', width=400, height=400, title=f"Eigenvalues {e1:.2f}, {e2:.2f}")

def cx_portrait( r, theta, v1,v2, N_iterates):
    bounds = (-3,3)
    pp.set( r*rotation_matrix_A(theta*np.pi/180))

    h = PhasePortrait.background(pp.dydx,bounds=bounds) +\
        plot_xy( *pp.iterates((v1,v2), N_iterates-1) )

    return hv.Overlay(h).opts(show_legend=True, legend_position='top', width=400, height=400, title=f"Eigenvalues r exp(i), r={r:.2f}, theta={theta:.2f}")
```

## 2.4 Eigenvalues and Eigenvectors of Positive Stochastic Matrices

A positive stochastic matrix is a square matrix with the following properties:
* All entries are positive (strictly greater than zero).
* The sum of each column is equal to 1.

This type of matrix has several **important characteristics:**
* It is a special case of a stochastic matrix, which only requires non-negative entries and column sums equal to 1.
* The Perron-Frobenius theorem applies to positive stochastic matrices, guaranteeing:
    * The eigenvalue 1 has algebraic and geometric multiplicity of 1.
    * There exists a unique eigenvector corresponding to eigenvalue 1 with all positive entries<br>and summing to 1 (called the **steady state vector**).
    * All other eigenvalues have absolute value strictly less than 1.

For any initial probability vector, repeated multiplication by a **positive stochastic matrix** will **converge to the steady state vector.**

Positive stochastic matrices are particularly useful in modeling Markov chains<br>
where there is a non-zero probability of transitioning between any two states in a single step.

#### Example

```python
import sympy as sp

def compute_2D_stationary_vector(s_vec):
    s_vec = s_vec/np.linalg.norm(s_vec)
    if np.any(s_vec < 0): s_vec = - s_vec
    return s_vec

A = np.array([[0.7, 0.2],
              [0.3, 0.8]])
side_by_side( " A =", sp.Matrix(A))
E     = np.linalg.eig(A)
s_vec = compute_2D_stationary_vector(E.eigenvectors[:,1])

print("Eigenvalues:      ", np.linalg.eigvals(A))
print("Stationary vector:", s_vec)
print()
print( "check the stationary vector has eigenvalue 1: || A s - s|| = 0: ", np.linalg.norm(A @ s_vec - s_vec) < 1e-12)
```

```python
pp= PhasePortrait(A)

c=3.55
h = PhasePortrait.background(pp.dydx,bounds=(-3,3)) +\
    plot_xy( *pp.iterates((3,0), 20) )+\
    [hv.Curve( ([-c*s_vec[0],c*s_vec[0]], [-c*s_vec[1], c*s_vec[1]])).opts(color="red", line_width=1 )]

print( "Iteration converges to the Stationary Vector line")
pn.Row(hv.Overlay(h).opts(show_legend=True, legend_position='top', width=400, height=400))
```

# 3. Take Away

**Phase Space Representation**

In the context of this linear process, the phase space is a multidimensional space where each dimension corresponds to a component of the state vector $x_n$. The dimensionality of the phase space is equal to the number of rows (or columns) in the matrix $A$.

**Key Characteristics:**

* **State Vector:** Each point in the phase space represents a possible state of the system, given by the vector $x_n$.
* **Trajectories:** As the system evolves over time (i.e., as n increases), it traces out a trajectory in the phase space. Each trajectory is a sequence of points representing the system's state at successive time steps.
* **Fixed Points:** Fixed points (also called **equilibrium points**) of the system are states where $x_n = x_{n-1}$. These points satisfy the equation $x = A x$, or equivalently, $(A - I) x = 0$, where $I$ is the identity matrix.

**Analysis of System Behavior**

The behavior of trajectories in the phase space is determined by the properties of the matrix $A$:
* **Eigenvalues and Eigenvectors:** The eigenvalues $\lambda$ and eigenvectors $v$ of $A$ play a crucial role in determining the system's behavior.
* **Stability:** The stability of the system is determined by the magnitude of the eigenvalues:
    * If all $\vert \lambda\vert < 1$,  the system is stable, and all trajectories converge to the origin.
    * If any $\vert \lambda\vert > 1$, the system is unstable, and some trajectories will diverge.
    * If the largest $\vert \lambda\vert = 1$, the system may exhibit periodic or quasi-periodic behavior.
* **Invariant Subspaces:** Eigenvectors of $A$ define invariant subspaces in the phase space. If $x_0$ is an eigenvector of $A$, then the entire trajectory will lie along the line defined by that eigenvector (subject to numerical error!)

**Long-term Behavior**

As $n$ approaches infinity, the long-term behavior of the system is determined by the dominant eigenvalue (the eigenvalue with the largest magnitude):
* If $\vert\lambda_{\text{dominant}}\vert < 1$, all trajectories converge to the origin.
* If $\vert\lambda_{\text{dominant}}\vert > 1$, trajectories diverge (except possibly along certain eigenvectors).
* If $\vert\lambda_{\text{dominant}}\vert = 1$, the behavior can be more complex, potentially resulting in periodic orbits or other interesting structures in the phase space.

# 1. Stochastic Matrix

## 1.1 Definitions

<div style="background-color:#F2F5A9;color:black;height:1cm;">

**Definition:**
A probability vector $v$ is a vector whose elements are nonnegative real numbers that sum to 1.
</div>

<div style="background-color:#F2F5A9;color:black;">

**Definition:**
* A **right stochastic matrix** is a square matrix of nonnegative real numbers, with each row summing to 1.
* A **left stochastic matrix** is a square matrix of nonnegative real numbers, with each column summing to 1.
* A **doubly stochastic matrix** is a square matrix of nonnegative real numbers with each row and column summing to 1.
</div>

We will use the convention that the columns are probability vectors, i.e. that the stochastic matrices are left stochastic matrices.

**Example:**
$\;\;P = \begin{pmatrix}
.1 & 0 & 0 & .2  \\
.3 & 1 & .9 & .4 \\
.2 & 0 & 0 &  .1 \\
.4 & 0 & .1 & .3
\end{pmatrix}$
 is a stochastic matrix.

**Remark:** By definition the vector of all ones is an eigenvector of a right stochastic matrix with eigenvalue $\lambda = 1$.

<div style="background-color:#F2F5A9;color:black;">

**Definition:** A **Markov chain** is a stochastic process describing a sequence of possible events<br>
$\qquad$ where the probability of each event depends only on the state attained in the previous event.

A **markov chain** It can be represented by:
* A state space $S$
* A transition probability matrix $P$ (a stochastic matrix)
* An initial probability distribution (a probability vector)
</div>

#### Markov Chain Example

Consider the following two examples of the probability of moving between socioeconomic classes.<br>
* There are three possible states: {Rich, Middle Class, Poor}
* The directed edges indicate the probability of transitioning from a given state to another state (not necessarily distinct!)
* For each state, the transition probabilities to another state (including the current state) sum to 1.

<br>The transition probability matrix $P = (p_{i j})$, where index $i$ is an ending state and $j$ is a starting state,<br>
with increasing indices $i$ and $j$ corresponding to each of the states {Rich, Middle Class, Poor} in turn.<br>
I.e., a row of the matrix records the probabilities of incoming transitions.

The first example has
$\;\;P_1 = \begin{pmatrix} 0.8 & 0.2 & 0.0 \\ 0.1 & 0.4 & 0.1 \\ 0.1 & 0.4 & 0.9 \end{pmatrix},\;\;$

while the second more pessimistic example has
$\;\;P_2 =\begin{pmatrix} 0.9 & 0.3 & 0.0 \\ 0.1 & 0.4 & 0.0 \\ 0.0 & 0.3 & 1.0 \end{pmatrix}.$

The matrices $P_1$ and $P_2$ **are (left) stochastic matrices** and express the transition probabilities from state to state over a given period of time.

Note that in the second model $P_2$, the poor remain poor forever!

<div style="background-color:#F2F5A9;color:black;height:1.0cm;">

**Definition:**
A state $i$ is **absorbing** if it is impossible to leave this state.
</div>

Therefore, the state $j$ is absorbing if $P_{j j} = 1$ and $P_{i j} = 0$ for $j \ne i$.<br>
If every state can reach an absorbing state, then the Markov chain is an absorbing Markov chain.

For $P_2$, "poor" is and absorbing state, and the corresponding Markov Chain is an absorbing chain.

We define a **probability vector** to be the state distribution, e.g., $v = \begin{pmatrix} 0.1 \\ 0.5 \\ 0.4 \end{pmatrix}$<br>
expresses that $10$% of a given population is rich, $50$% is middle class, and the remaining $40$% of the population is poor.

The product of a a stochastic matrix with a probability vector yields another probability vector,<br>
representing the state distribution after one transition in the Markov chain:<br>
a given percentage of rich people loose money and join the middle class or become poor, etc.

In our example $\tilde{v} = P_1 v = \begin{pmatrix} 0.17 \\ 0.38 \\ 0.41\end{pmatrix},$ i.e., a new distribution of wealth.

**Remark:** the convention of using left stochasic matrices, i.e., $P_1^t$,<br>
would multiply the matrices by a row vector from the left: $\;\;\tilde{v}^t = v^t P_1$

```python
P_1 = np.array([[0.8,0.2,0.0],[0.1,0.4,0.1], [0.1,0.4,0.9]])
P_2 = np.array([[0.9, 0.3, 0.0],[0.1,0.4,0.0],[0.0,0.3,1.0]])
v   = np.array([0.1,0.5,0.4])
np.array([P_1 @ v, P_2 @ v])
```

Thus the transition from a current probability vector $v_0$ to a new probability vector $v_1 = P v_0$ is a state transition in a Markov chain:<br>
$\qquad v_1 = P v_0$.

The markov chain assumption keeps the transition matrix $P$ constant: we can apply it repeatedly to observe the evolution of a probability vector:<br>
$\qquad v_n = P v_{n-1}, n=1, 2, \dots$ for a given initial probability $v_0$.

(See [**DifferenceEquations.ipynb**](DifferenceEquations.ipynb))

## 1.2 The Evolution of the Probability Vector

We can observe the evolution of the probability vector for a given Markov chain.<br>
Continuing our previous examples
with the stocastic matrices $P_1$ and $P_2$ we see the following:

```python
def chain(P, v, n, state_names):
    results = []
    results.append(v.copy())
    for _ in range(n):
        v = P @ v
        results.append(v.copy())
    return pd.DataFrame( np.array(results), columns=state_names)

df_1 = chain(P_1, v, 100, ['rich', 'middle class', 'poor'])
df_2 = chain(P_2, v, 100, ['rich', 'middle class', 'poor'])
```

```python
h_1 =\
hv.Curve( df_1, 'index', 'rich',         label='rich').opts(title='Transition Matrix P_1', show_grid=True) *\
hv.Curve( df_1, 'index', 'middle class', label='middle class')*\
hv.Curve( df_1, 'index', 'poor',         label='poor')
h_2 =\
hv.Curve( df_2, 'index', 'rich',         label='rich').opts( title='Transition Matrix P_2', show_grid=True)*\
hv.Curve( df_2, 'index', 'middle class', label='middle class')*\
hv.Curve( df_2, 'index', 'poor',         label='poor')

pn.Row((h_1 + h_2).opts("Overlay", legend_position='bottom'))
```

We see that for $P_2$, the absorbing state "poor" eventually accumulates all the probability: $P_2$ has an eigenpair $(1,\begin{pmatrix}0\\0\\1\end{pmatrix})$.

A probability vector that is an eigenvector of a transition matrix with eigenvalue 1 is called a **stationary distribution.**<br>
$\qquad$ It can be shown that all stochastic matrices have a stationary distribution:<br>
$\qquad$ Since $\lambda=1$ is always an eigenvalue, the corresponding Eigenspace basis must have at least one eigenvector.<br>
$\qquad$ The Perron-Frobenius Theorem ensures that one can always choose this basis vector to be a probability vector.

```python
#check: P_2 v = v  iff  || P_2 v -v || = 0
np.linalg.norm(P_2@np.array([0,0,1]) - np.array([0,0,1])) == 0
```

## 1.3 Phase Curves

Stochastic matrices are not all diagonalizable. When they are, however we can write the probability vector iterates in terms of an eigenvector expansion of the starting value.

Let $A$ be diagonalizable with eigenpairs $(\lambda_1, s_1), (\lambda_2, s_2), \dots, (\lambda_k, s_k)$ 

Express the starting probability distribution $v_n$ in this basis: $v_0 = \alpha_1 s_1 + \alpha_2 s_2 + \dots + \alpha_k s_k$. Then<br>
$\qquad v_n = A^n v_0 & = \alpha_1 \lambda_1^n s_1 + \alpha_2 \lambda_2^n s_2 + \dots + \alpha_k \lambda_k^n s_k 
$

We now define a function $f(t)$ setting $n = t$ and allowing $t$ to take on any value, i.e., <br>
$\qquad f(t) = \alpha_1 \lambda_1^t s_1 + \alpha_2 \lambda_2^t s_2 + \dots + \alpha_k \lambda_k^t s_k$.<br>
The resulting curve $x = f(t)$ is know as a **phase curve** and has the property that each point of the Markov chain starting with initial condition $v_0$ lies on this curve.

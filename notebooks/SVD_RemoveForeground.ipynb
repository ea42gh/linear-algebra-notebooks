# 1. Utility Functions and Data Sets

## 1.1 Display Images, Convert to and from Matrix of Vectorized Images

```python
def raster(img):  return hv.Raster(img).opts(cmap="gray", xaxis=None, yaxis=None, frame_width=200, aspect='equal')

def img_to_vec( img ) :               return img.reshape(1, -1)
def vec_to_img( vec, height, width ): return vec.reshape(height,width)
```

## 1.2 Load a Sequence of Image Frames

```python
# Load A sequence of Image Frames
def get_data_Matrix (mypath="movies/peds"):
    cwd = getcwd()

    mypath = join(cwd,mypath)
    files = [join(mypath,f) for f in listdir(mypath) if isfile(join(mypath, f)) and f.startswith(".")==False]
    # Read image
    img = Image.open(files[0])
    I = np.array(img)
    # Output Images

    Height, Width = I.shape

    X = np.zeros((len(files),Height*Width))
    for i,file in enumerate(files):
        img = Image.open(file)
        I = np.array(img)
        X[i, :] = I.reshape(1, -1)
    return X, Height, Width
```

```python
X, Height, Width = get_data_Matrix(mypath="movies/peds")
NumFrames = X.shape[0]
print( f" Frames: {NumFrames},    Image Size: {Height, Width}" )
```

# 2. Process the Data

## 2.1 A sequence of Images

```python
interact( lambda frame: raster( vec_to_img(X[frame-1,:], Height, Width )).opts(frame_width=400), frame = range(1,NumFrames+1))
```

We see a sequence of Frames showing people walking along a path. The background is fairly constant.

If we vectorize the frames and combine them in a matrix as columns,<br>
we expect that the rank 1 approximation of the matrix using the SVD will show an average image:<br>
$\qquad$ i.e., it should remove the variations in the frame (the people walking!)

## 2.2 The SVD of the Data Matrix

```python
U, , VT = svd(X, full_matrices=False)
S = np.diag()
```

```python
h_ = hv.Curve( , "index", "" ).opts(logy=True, width=500, title="Distribution of the Singular Values")

h_percent_variance = hv.Curve(100*np.cumsum(**2) / (**2).sum(), "index", "% Variance Explained")\
                       .opts(logy=True, width=500, title="Percentage of Variance Explained")
pn.Row( h_, h_percent_variance)
```

As expected, the first frame accounts for 98.5% of the Variance!

# 2.3 Reconstruction of the Frames

Let's take a look: we will use a rank $k$ approximation of the Data Matrix, and look at individual frames:
* the rank 1 approximation should show the backgound. Higher rank approximations will reintroduce the people walking
* different frames will reconstruct the individual frames in our image

```python
# rank = k approximation
def reconstruct( U, S, VT, k, i ):
    Xhat = U[:, :k]@S[0:k, 0:k]@VT[:k, :]
    return vec_to_img( Xhat[i,:], Height, Width )

interact( lambda k, frame: raster(reconstruct(U,S,VT,k,frame-1)).opts(frame_width=400),
          k=range(1,NumFrames+1), frame=range(1,NumFrames+1))
```

___
Can we remove the background and instead see the people walking?

Lets try reconstructing the frames by using the Singular values starting at index $k$

```python
def drop_initial_sigmas( U, S, VT, k, i ):
    Xhat = U[:, k:]@S[k:, k::]@VT[k:, :]
    return vec_to_img( Xhat[i,:], Height, Width )

interact( lambda k, frame: raster(drop_initial_sigmas(U,S,VT,k,frame-1)).opts(frame_width=400),
          k=range(2,NumFrames), frame=range(1,NumFrames))
```

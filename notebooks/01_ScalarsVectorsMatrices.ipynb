# 1. Introduction

## 1.1 Scalars

**Scalars** are numbers, with addition, subtraction, multiplication and division

<div style="background-color:#F2F5A9;color:black;">

| Scalar Type | Symbol | Examples |
| ---         | ---    | ---      |
| **Integers** | $\mathbb{Z}$ |$ 3,\ -5,\ 0,\ 13418$ |
| **Rationals** | $\mathbb{Q}$ | $\frac{3}{5},\ -\frac{7}{2},\ -9$ |
| **Reals**    | $\mathbb{R}$ | $$3.14,\ \sqrt{2},\ -\frac{6}{17},\ e^\pi$$ |
| **Complex Numbers** | $\mathbb{C}$ | $8-7i,\ 7\pi,\ 4i$ |
| **Integers Mod 2** | $\mathbb{Z}_2$ | $0,\ 1$ |

<br>
</div>

**Remarks**
* **Closure:** a set is **closed** under an operation if performance of that operation on members of the set always produces a member of that set, e.g.,
    * integer + integer -> integer
    * rational - rational -> rational
* the **integers are not closed under division:** $3 / 5$ is not a integer
* we will use $\color{darkred}{\mathbb{F} = \mathbb{Q}, \mathbb{R}, \mathbb{C}, \mathbb{Z}_2}$

* <strong  style="color:darkred;size:40px;">most of our examples will use $\mathbb{Q}$</strong> $\quad$ (exact arithmetic!)
* using reals $\mathbb{R}$ requires approximations, since computations are carried out with finite precision

## 1.2 Vectors and Matrices

<div style="background-color:#F2F5A9;color:black;">

**Definition:** **Vectors** are an **ordered set** of scalars.<br>
$\qquad$ Each of the scalars is an **entry** of the vector.<br>
$\qquad$ The position of the entry in the ordered set is the **index** of the entry.
</div>

> **Example:** $( 1,\ \; 3,\ \; -5)$ is a vector consisting of 3 scalars<br>
$\quad$ first entry 1, second entry 3, third entry -5

##### **Notation:**

* we will write $\color{darkred}{v = (v_1 \; v_2 \; \dots v_n)}$ for a vector $v$ with $n$ scalar **entries** $\; v_1, v_2, \dots$ 
* a vector $v$ with $n$ entries in the scalars $\mathbb{F}$<br>$\quad\quad$ is denoted by $\quad$ $\color{darkred}{v \in \mathbb{F}^n}$
* the number of entries $n$ is the **size** of the vector

---
<div style="background-color:#F2F5A9;color:black;height:2cm;">

**Definition:** A **matrix** of **size** $M \times N$ is a rectangular array of scalars in $\mathbb{F}$<br>
    $\qquad$ called **entries** of the matrix<br>
    $\qquad$ arranged in $M$ rows by $N$ columns.<br>    
</div>

> **Example:**
$$\begin{pmatrix} 3 & 5 & -2 \\ 4 & 8 & 3 \end{pmatrix}$$ 

##### **Notation:**

* We will write $\color{darkred}{A = \begin{pmatrix} a_{1 1} & a_{1 2} & \dots & a_{1 N} \\
                                     a_{2 1} & a_{2 2} & \dots & a_{2 N} \\
                                     \dots & \dots & \dots & \dots \\
                                     a_{M 1} & a_{M 2} & \dots & a_{M N} \\                                    
    \end{pmatrix}}$ $\quad$ for a matrix $A$ with **entries** listed in $M$ rows and $N$ columns

* a matrix $A$ with $M \times N$ entries in the scalars $\mathbb{F}$<br>$\quad\quad$ is denoted by $\quad$ $\color{darkred}{A \in \mathbb{F}^{M \times N}}$
* the number of rows and columns $M \times N$ is the **size** of the matrix

**Remarks**
* **vectors** require a **single index**
* **matrices** require **2 indices**: a **row index** $i$ and a **column index** $j,\;$ where $A = \left( a_{i j} \right)$

* **vectors** can be written as **either a row or a column**
* **matrices** have a **layout convention:** the first index $i$ denotes the row, the second index $j$ denotes the column

##### **Notation convention:**

* scalars will typically be denoted by **greek letters** or by the matrix/vector name with appropriate indices: $\color{darkred}{\alpha,\; a_4,\; a_{2 3}}$
* vectors will typically be denoted by **lower case latin alphabet letters:** $\quad \color{darkred}{u,\; v, \; x}$
* matrices  will typically be denoted by **upper case latin alphabet letters:** $\; \color{darkred}{A,\; B,\; X}$

# 2. Special Cases

## 2.1 Matrices With a Single Row or a Single Column

<div style="background-color:#F2F5A9;color:black;">

**Definition:**
* A **row vector** is a matrix of size $1 \times N$
* A **column vector** is a matrix of size $M \times 1$
</div>

##### **Example:**

> * $A = \left(  3 \; 9 \; 5 \right)\;$ is a row vector in $\mathbb{Q}^{1 \times 3}$ $\quad$ (a **matrix** of size $1 \times 3:$
$\quad A = \left( a_{\color{darkred}1 1}\;a_{\color{darkred}1 2}\; a_{\color{darkred}1 3} \right)\ $ )

> * $A = \begin{pmatrix}  3 \\ 9 \\ 5 \end{pmatrix}\;$ is a column vector in $\mathbb{Q}^{3 \times 1}$ $\quad$ (a **matrix** of size $3 \times 1:$ $\quad A = \begin{pmatrix} a_{1 \color{darkred}1}\\ a_{2 \color{darkred}1}\\ a_{3 \color{darkred}1} \end{pmatrix}\ $ ) 

> * $v = \begin{pmatrix}  3, & 9, & 5 \end{pmatrix}\;$ is a vector in $\mathbb{Q}^{3}$ $\quad$ (a **vector** of size $3:$ $\quad v = \left( v_1\; v_2 \; v_3 \right)\;$ ) 

**Remarks:**
* **row vectors and column vectors have a layout convention**; vectors do not
* we will (silently!) **convert** row and/or column vectors to vectors **by dropping an index**
* we will (silently!) **convert** vectors to row vectors or column vectors by **adding a fixed column or row index**
* Typically, **vectors will be converted to column vectors**

* We can **convert a matrix of size $1 \times 1$** to a vector of size $1$ to a scalar<br>
> Example: $A = \left( a_{1 1} \right) \longleftrightarrow a = \left( a_1 \right) \longleftrightarrow a:$ $\quad ( 5 ) \longleftrightarrow 5$

## 2.2 Zero Matrix, Identity Matrix

Two special matrices that we will encounter frequently are
* a **zero matrix** with all entries equal to the scalar 0: $$0 = \begin{pmatrix} 0 & 0 & 0 \\ 0 & 0 & 0 \end{pmatrix}$$<br>

* an **identity matrix,** i.e., a **square matrix** (size $N \times N$) with entries $1$ on the **main diagonal** (entries $a_{i i}$), and zeros everywhere else<br><br>
$$I = \begin{pmatrix} \color{darkred}1 & 0 & 0 \\ 0 & \color{darkred}1 & 0\\ 0 & 0 & \color{darkred}1 \end{pmatrix}$$

#### **Remark:**

* the size of the zero matrix or the identity matrix will usually be clear from the context
* we will denote them with $0$ and $I,$ regardless of their actual size
* if necessary, size will be indicated by a subscript: $0_{2 \times 3}$ and $I_3$ or $I_{3 \times 3}$

# 3. Geometrical Representation of Vectors

## 3.1 Arrows

Euclidean vectors in $2D$ and $3D$ can be represented as arrows (e.g., directed quatities in physics: velocity, force, ...)

While we cannot represent vectors with more than 3 entries directly,<br>$\quad$  we can **use our geometrical intuition** and continue to think of vectors as<br>$\quad$  **arrows in a higher dimensional space.**

## 3.2 Functions

Since the entries in a vector are ordered, we can think of $v = \left( v_1, v_2, \dots v_n \right)$ as a function from the index $i$ to the entry $v_i$:
$$i \in \{ 1, 2 \dots n \} \rightarrow v_i$$

```julia
x= [1 2 3 4 5];y=[2 4 1 3 2]
scatter( x,y, color=:red, legend=:none)
bar!(x,y, bar_width=0.01, color=:blue, xlim=(0.8,5.2), ylim=(0,4.2))
plot!(title = "vector v=$y", size=(400,200))
```

## 3.2 Positive Definite Matrix Tests

Remark: If $A$ is negative (semi)definite, then -A is positive (semi)definite

Equivalent Positive Definite Tests
* $x^t A x > 0$ for all $x \ne 0$
* All eigenvalues $\lambda > 0$
* Leading principal minors are all positive (Sylvester's criterion)
    * For positive Semi-definite, we would need to check ALL minors are non-negative!
* Every pivot in a row echelon form obtained without scaling and/or row interchanges) is positive, i.e.,
    * A has an LU decomposition $A = L U$, such that all pivots in $U$ are positive, with no missing pivots ($A$ is invertible)
* $A = R^t R$ where $R$ has independent columns
    (note many choices for $R$

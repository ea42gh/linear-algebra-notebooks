# 1. The Similarity Transform

## 1.1 Change of Basis

**Change coordinate systems** (an application of linear transformations) 

<div>
<div style="float:left;"><img src="SalzburgBasis.png" width=400>
    
The two representations are related by
$$
x_p = \tilde{x}_1 \tilde{s}_1 + \tilde{x}_2 \tilde{s}_2
$$
</div>
<div style="float:center;">&ensp;&ensp;</div>
<div style="float:right;">Given a point $p$ with coordinates $x_p = \begin{pmatrix} x_1 \\ x_2 \end{pmatrix}$<br>
in the original coordinate system.

Given new coordinate vectors $\{ \tilde{s}_1, \tilde{s}_2 \}$,<br>
we can describe $p$<br>
by its new coordinate vector $\tilde{x}_p = \begin{pmatrix} \tilde{x}_1 \\ \tilde{x}_2 \end{pmatrix}$.
</div></div>

---
This is an $A x = b$ type of problem in column view:
$$
x_p = \tilde{x}_1 \tilde{s}_1 + \tilde{x}_2 \tilde{s}_2
\quad\Leftrightarrow\quad
S \begin{pmatrix} \tilde{x}_1 \\ \tilde{x}_2 \end{pmatrix}
= \begin{pmatrix} x_1 \\ x_2 \end{pmatrix}, \quad \text{ where } S = \begin{pmatrix} \tilde{s}_1 & \tilde{s}_2 \end{pmatrix}.
$$

**Remark:**
* Since the coordinate vectors are chosen to be linearly independent, $S$ is invertible.


* **<span style="color:red;">Switching coordinate systems $x_p \rightarrow \tilde{x}_p$ is accomplished by a substitution:</span>**
$$\mathbf{x_p = S\ \tilde{x}_p \; \Leftrightarrow \; \tilde{x}_p  = S^{-1} x_p}.$$
* The columns of $S$ are the new coordinate vectors expressed in the original coordinate system.

---

#### Example:

<div style="float:left;"><img src="ChangeOfBasis.svg" width=250></div>
<div style="float:right;">
In the original coordinate system, $x_p = \begin{pmatrix} x_1 \\ x_2\end{pmatrix} = \begin{pmatrix} 1 \\ 6\end{pmatrix}$.<br>
In this coordinate system, the new coordinate vectors are $$\tilde{a}_1 = \begin{pmatrix} 3\\2 \end{pmatrix},\quad
\tilde{a}_2 = \begin{pmatrix}-3\\3 \end{pmatrix}$$.

The coordinates of the point in the new coordinate system is the solution of
$$
\begin{pmatrix} 3 & -3 \\ 2 & 3 \end{pmatrix}
\begin{pmatrix} \tilde{x}_1 \\ \tilde{x}_2 \end{pmatrix}
= \begin{pmatrix} 1 \\ 6 \end{pmatrix}.
$$

For  any point $x_p$, we have
$$
\begin{pmatrix} 3 & -3 \\ 2 & 3 \end{pmatrix}
\begin{pmatrix} \tilde{x}_1 \\ \tilde{x}_2 \end{pmatrix}
= \begin{pmatrix} x_1 \\ x_2 \end{pmatrix} \; \Leftrightarrow 
\left\{
\begin{align}
x_1 &= 3 \tilde{x}_1 - 3 \tilde{x}_2 \\
x_2 &= 2 \tilde{x}_1 + 3 \tilde{x}_2 \\
\end{align}
\right. .
$$
</div>

## 1.2 Similarity Transform

### 1.2.1 Definition

Let $A$ be a square matrix of size $N \times N$.<br>
The linear transformation $y = A x$ transforms $x \in \mathscr{R}^N \longrightarrow \mathscr{R}^N$.<br>
Since the domain and codomain are both $\mathbb{R}^N$, we can introduce the same change of basis:
$x = S \tilde{x}, y = S \tilde{y}$.

Substitution in $y = A x$ yields
$$
S \tilde{y} = A S \tilde{x}  \Leftrightarrow \tilde{y} = S^{-1} A S \tilde{x}.
$$

The linear transformation expressed in the new coordinate system is $\tilde{A} = S^{-1} A S$.

<div style="background-color:#F2F5A9">

**Definition:** Two matrices $A$ and $B$ of same size $N \times N$ are **similar** iff
there exists an invertible matrix $S$ such that
$$ B = S^{-1} A S. $$
 
$\quad\;$ The transformation $\tilde{A} = S^{-1} A S$ is a **similarity transformation.**
<div>

#### Example

Consider the projection onto the $x_1$ axis:
$$
\begin{pmatrix} y_1 \\ y_2 \end{pmatrix} = A \begin{pmatrix} x_1 \\ x_2 \end{pmatrix},\quad\text{ where }\;
A =\begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix}.
$$

Let us use a set of new axes rotated by $\theta=\frac{\pi}{4}$ radians compared to the original.<br>

The similarity transform recovers the familiar result of a projection onto $\tilde{x}_2 = \tilde{x}_1$:
$$
R = \frac{\sqrt{2}}{2}\begin{pmatrix} 1 & - 1 \\ 1 & 1 \end{pmatrix}, \quad \tilde{P} = R^{-1} P R = 
\frac{\sqrt{2}}{2}\begin{pmatrix} 1 &  1 \\ -1 & 1 \end{pmatrix}
\begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix}
\frac{\sqrt{2}}{2}\begin{pmatrix} 1 & - 1 \\ 1 & 1 \end{pmatrix}
= \frac{1}{2} \begin{pmatrix} 1 & -1 \\ -1 & 1 \end{pmatrix}.
$$

### 1.2.2 Similar Matrices and Eigenpairs

<div style="background-color:#F2F5A9">

**Theorem:** Let $A$ be a square matrix with an eigenpair $(\lambda, x)$.<br>
$\quad\;$ Any similar matrix $\tilde{A} = S^{-1} A S$ has eigenpair $(\lambda, S^{-1} x)$.
<div>

Geometrically, this seems obvious: since similar matrices describe the same geometrical transformation,<br>
they should have the same eigenvalues with corresponding eigenvectors:<br>
$\tilde{x} = S^{-1} x$ is the eigenvector $x$ expressed in the new coordinate system.

Algebraically, we see that the substitution $x = S \tilde{x}$ yields
$$
\begin{align}
A x = \lambda x & \Leftrightarrow A\ S \tilde{x}  = \lambda S \tilde{x} \\
& \Leftrightarrow S^{-1} A S\ \tilde{x} = \lambda \tilde{x} \\
& \Leftrightarrow \tilde{A} \tilde{x} = \lambda \tilde{x}, \\
\end{align}
$$
where we have set $\tilde{A} = S^{-1} A S$ and $\tilde{x} = S^{-1} x$.

## 1.3 Special Case: A Basis of Eigenvectors

### 1.3.1 A $2 \times 2$ Example

Let us look at a simple $2 \times 2$ example:
$$
A = \begin{pmatrix} 6 & 10 \\ -2 & -3 \end{pmatrix}.
$$

#### **Step 1:** Find the roots of the characteristic polynomial

$$
\begin{align}
p(\lambda) &= det ( A - \lambda I ) \\
&=
\begin{vmatrix}  6 -\lambda & 10 \\ -2 & -3 -\lambda \end{vmatrix} \\
&= \lambda^2 - 3 \lambda + 2 \\
&= (\lambda - 1) (\lambda -2),\quad\text{ eigenvalues } \lambda=1,2
\end{align}
$$

#### **Step 3:** Similarity transform: $A$ in the Eigenvector Basis

Since the eigenvectors are linearly independent, we can use them as a new basis for $y = A x$.

Set $S = \begin{pmatrix} -2 & 1 \\ -5 & 2 \end{pmatrix}$ and compute
$$
\begin{align}
\tilde{A} & = S^{-1} A S \\
          & = \begin{pmatrix} 2 & -1 \\ 5 & -2 \end{pmatrix}
              \begin{pmatrix} 6 & 10 \\ -2 & -3 \end{pmatrix}
              \begin{pmatrix} -2 & 1 \\ -5 & 2 \end{pmatrix} \\
          & = \begin{pmatrix} 1 & 0 \\ 0 & 2 \end{pmatrix}, \quad\quad\quad\quad \color{red}{\text{ A diagonal matrix!}}
\end{align}
$$

### 1.3.2 **Should we have predicted this?**

#### 1.3.2.1 **Linearly Independent Basis of Eigenvectors**

<div style="background-color:#F2F5A9">
    
**Theorem:** Eigenvectors for different eigenvalues are linearly independent.<br><br>

</div>

Given eigenpairs $(\lambda_1, x_1)$ and $(\lambda_2, x_2)$ of a square matrix $A$ with $\lambda_1 \ne \lambda_2$,<br>
Let's check linear independence:
$$
\alpha_1 x_1 + \alpha_2 x_2 = 0 \label{eq1}\tag{1}
$$

$
\begin{align}
&\; \text{Multiply Eq 1 from the left by } A : \quad&  \alpha_1 A x_1 + \alpha_2 A x_2 = 0 & \Rightarrow\quad &\ \alpha_1 \lambda_1 x_1 + \alpha_2 \lambda_2 x_2 = 0. &\\
&\ \text{Multiply Eq 1 by } \lambda_2 : \quad && \Rightarrow\quad & \alpha_1 \lambda_2 x_1 + \alpha_2 \lambda_2 x_2 = 0. & \\
&\text{Subtracting the resultant equations } : \quad && \therefore\quad & \alpha_1 (\lambda_2-\lambda_1) x_1  = 0. & \\
\end{align}
$

> **Remark:** Since we choose linearly independent vectors for the null space $\mathscr{N}(A-I)$,<br> $\quad\quad$ **the eigenvectors in our summary table will always be linearly independent.**

#### 1.3.2.2 **The Matrix $A$ Expressed in the Eigenvector Basis is Diagonal**

Let's look at the matrix of eigenvectors  $\mathbf{S = ( x_1 \, x_2 \, \dots x_k ) }$:

$$
\begin{align}
A x_i = \lambda_i x_i,\; i=1,2,\dots k & \; \Leftrightarrow \; &
    A (x_1 \; x_2 \; \dots \; x_k ) & = \ ( \lambda_1 x_1 \; \lambda_2 x_2 \; \dots \lambda_k x_k ) \\
    & & & = \ ( x_1 \; x_2 \; \dots \; x_k ) \begin{pmatrix} \lambda_1 & 0         & \dots  & 0 \\
                                                                     0 & \lambda_2 & \dots  & 0 \\
                                                                 \dots & \dots     & \ddots & \dots  \\                                
                                                                     0 & 0 & \dots  & \lambda_k \\
    \end{pmatrix}
\end{align}
$$

i.e., if we set $S = ( x_1 \; x_2 \; \dots x_k )$, we find $\mathbf{\color{red}{ A S = S \Lambda }}$, where
$$
\Lambda = \begin{pmatrix} \lambda_1 & 0         & \dots  & 0 \\
                                                                     0 & \lambda_2 & \dots  & 0 \\
                                                                 \dots & \dots     & \ddots & \dots  \\                                
                                                                     0 & 0 & \dots  & \lambda_k \\
    \end{pmatrix}
$$

is the diagonal matrix of the eigenvalues corresponding to each eigenvector.

##### **Do we have enough vectors?**

The eigenvectors are linearly independent, so $S$ has a pivot in every column. **To be invertible, it must be square.**

The $2 \times 2$ example we just did had 2 eigenvectors in the basis, so the $S$ matrix was invertible.

$\quad\quad\quad$ **This is not always the case!**

###### Example

Look at $A = \begin{pmatrix} 10 & 1 \\ 0 & 10 \end{pmatrix}.\quad $ **Step 1** results in $p(\lambda) = (\lambda-10)^2, $ so $\lambda=10$ is an eigenvalue with algebraic multiplicity 2.

**Step 2:** $A - 10 I =\begin{pmatrix} 0 & 1 \\ 0 & 0 \end{pmatrix}$ has $dim \mathscr{N}(A-10 I ) = 1 :$ the homogeneous solution is $x_h = \alpha \begin{pmatrix} 1\\0 \end{pmatrix}.$

# 2. Diagonalization

## 2.1 Basic Theory

### 2.1.1 Diagonalizable Matrices

##### **Definitions and Theorems**

<div style="background-color:#F2F5A9">

**Theorem:** Let $A$ be a matrix $A$ of size $N \times N$ with $N$
    linearly independent eigenvectors $x_i,\ i=1,2, \dots N$<br>
    $\quad\quad$ and corresponding (not necessarily distinct eigenvalues $x_i,\ i=1,2, \dots N.$<br><br>
$\quad\quad$ Let $S$ be the matrix with eigenvector columns $S = \begin{pmatrix} x_1 \ x_2 \ \dots \ x_N \end{pmatrix}$,<br>
     $\quad\quad$ and let
$$
\Lambda = \begin{pmatrix} \lambda_1 & 0         & \dots  & 0 \\
                                                                     0 & \lambda_2 & \dots  & 0 \\
                                                                 \dots & \dots     & \ddots & \dots  \\                                
                                                                     0 & 0 & \dots  & \lambda_k \\
    \end{pmatrix}
$$
be the diagonal matrix with the corresponding eigenvalues on the diagonal.

$$
\text{Then }\quad\mathbf{  A S = S \Lambda \; \Leftrightarrow S^{-1} A S = \Lambda \; \Leftrightarrow A = S \Lambda S^{-1} }.
$$
</div>

<div style="background-color:#F2F5A9">
    
**Definition:** A square matrix $A$ is **diagonalizable** (or **non-defective**)<br>
    $\quad\quad$ if there exists an invertible matrix $S$<br>
    $\quad\quad$ such that $S^{-1} A S$ is diagonal. Otherwise, the matrix is **defective.**
    
**Theorem:** Let $S$ be an invertible matrix that diagonalizes a square matrix $A.$<br>
    $\quad\quad$ The columns of $S$ are linearly independent eigenvectors of the matrix $A$.

</div>

**Remarks:**
* We have already seen one example where $A$ did not have enough linearly independent eigenvectors.
* To be diagonalizable, a square matrix $A$ of size $N \times N$ must have $N$ linearly independent eigenvectors.
* Since the eigenvectors are in $\mathbb{R}^n$, **the eigenvectors of a diagonalizable matrix form a basis for $\mathbb{R}^N$**.

##### **Why Algebraic Multiplicities are Important**

<div style="background-color:#F2F5A9">

**Theorem:** Let $A$ be a square matrix of size $N \times N$, and let $\lambda$ be an eigenvalue of $A$ with algebraic multiplicity $m_a$.<br>
$\quad\quad$ Let $E_\lambda = \mathscr{N}(A - \lambda I)$. Then
$$
1 \le dim E_\lambda \le m_a.
$$
</div>

**Remark:** The matrix **$A$ is diagonalizable iff $dim E_\lambda = m_a$ for all eigenvalues $\lambda$**.

<div style="background-color:#F2F5A9">

**Definition:** The **geometric multiplicity** $m_g$ of an eigenvalue $\lambda$ of a matrix $A$ is
    $$ m_g = dim \mathscr{N}(A - \lambda I) $$
    
**Theorem:** A matrix $A$ is diagonalizable iff the geometric maultiplicity $m_g$ and algebraic multiplicity $m_a$
    satisfy $$m_g = m_a$$ for all eigenvalues of $A$.
</div>

### 2.2.2 When is a Matrix Diagonalizable?

#### 2.2.2.1 Existence: No Repeated Eigenvalues

The simplest case is a matrix that has distinct eigenvalues:<br>
Since all eigenvalues have algebraic multiplicity $m_a = 1$,
we will have a complete basis.

**Example:**
$$ A = \begin{pmatrix} 2 & 1 \\ 0 & 3 \end{pmatrix} $$

has distinct eigenvalues $\lambda = 2, 3$ with corresponding eigenvectors
$$ \begin{pmatrix} 1 \\ 0\end{pmatrix}, \; \text{ and } \begin{pmatrix} 1 \\ 1 \end{pmatrix} \quad\text{respectively.} $$

<div style="background-color:#F2F5A9">

**Theorem:** A square matrix $A$ of size $N \times N$ with $N$ distinct eigenvalues<br>
    $\quad\quad$ is diagonalizable.
</div>

#### 2.2.2.2 Existence: Symmetric Matrices

Symmetric matrices are quite special.<br>
We will extend the following theorem with further properties later.<br><br>

Here is the preliminary version:
<div style="background-color:#F2F5A9">

**Theorem:** A symmetric matrix $A$ with real coefficients
* is diagonalizable
* the eigenvalues and eigenvectors are real
</div>

#### 2.2.2.3 Existence: Normal Matrices

The symmetric matrix theorem above is a special case of the following more general theorem:

<div style="background-color:#F2F5A9">

**Theorem:** A matrix $A$ satisfying $A A^t = A^t A$ is diagonalizable.

**Definition:** A matrix is normal iff $A A^t = A^t A.

**Remark:** this theorem (and hence the definition) generalizes to matrices with complex entries.
</div>

Examples are
* symmetric matrices:  $\quad\quad\quad A^t = A$
* skew symmetric matrices: $\quad A^t = - A$
* matrices such that $\quad\quad\quad\; A^{-1} = A^t\quad$    (these are **orthogonal matrices** which we will study later)

# 4. Applications

## 4.1 Powers of a Diagonalizable Matrix

##### **Powers of a diagonal matrix**

<div style="background-color:#F2F5A9">

$$

##### **Powers of a Diagonalizable Matrix**

Consider a diagonalizable matrix $A = S^{-1} \Lambda S .$

* $A^2 = ( S^{-1} \Lambda S ) (S^{-1} \Lambda S ) = S^{-1} \Lambda^2 S$

* $A^p = ( S^{-1} \Lambda S ) (S^{-1} \Lambda S ) \dots  (S^{-1} \Lambda S ) = S^{-1} \Lambda^p S,\; \text{where } p=1,2,\dots .$.

While defined for positive integer powers, this formula can be generalized.

<div style="background-color:#F2F5A9">

* $A^p = ( S^{-1} \Lambda S ) (S^{-1} \Lambda S ) \dots  (S^{-1} \Lambda S ) = S^{-1} \Lambda^p S$ for all $p > 0.$
* $A^0 = I$ by definition.
* $A^p = ( S^{-1} \Lambda S ) (S^{-1} \Lambda S ) \dots  (S^{-1} \Lambda S ) = S^{-1} \Lambda^p S$ for all $p < 0$ provided $A^{-1} exists.
</div>

The matrix inverse in particular follows from the observation that
$$A = S^{-1} \Lambda S \Rightarrow A^{-1} = S^{-1} \Lambda^{-1} S.$$

# 5. Take Away

The investigation of $y = A x$ for square matrices $A$ showed that
* the change of basis $y = S \tilde{y}, x=\tilde{x}$ results in<br>
$\tilde{y} = \tilde{A} \tilde{x}$, where the similar matrix $\tilde{A} = S^{-1} A S$<br>
represents the same geometrical transformation in the new basis.

* If $A$ has **a complete eigenvector basis** $n_1, x_2, \dots x_n$ in $\mathbb{R}^n$
    * $\tilde{A}$ is **diagonal** in this basis: $\tilde{A} = \Lambda$
    * the diagonal entries of $\Lambda$ are the **eigenvalues** of the corresponding eigenvectors

<div style="float:left;">

* Let $x = \alpha_1 x_1 + \alpha_2 x_2 + \dots \alpha_n x_n$ be a vector expressed in the eigenvector basis<br>
$\begin{align}
A x & = \alpha_1 A x_1         & + \alpha_2 A x_2           & + \dots & + \alpha_n A x_n \\
    & = \lambda_1 \alpha_1 x_1 & + \lambda_2 \alpha_2 A x_2 & + \dots & + \lambda_n \alpha_n A x_n, 
\end{align}$ i.e.,<br>
<br>**Applying $A$ to a vector scales each eigenvector component by the corresponding eigenvalue.**
    
* If the eigenvalue is **complex, the vector component is scaled and rotated.**
</div>
